 # <font color=brown> Spojitosť funkcie</font>

Existuje **niekoľko intuitívnych pohľadov na spojitosť funkcie**. Zosumarizujme si ich ešte pred tým, než vyslovíme exaktnú definíciu spojitosti. Všimnime si ale, že budeme rozlišovať spojitosť funkcie v bode a spojitosť funkcie na (uzavretom) intervale. Spojitosť funkcie na (uzavretom) intervale budeme neskôr definovať pomocou spojitosti funkcie v bode. 

* Zhruba povedané, funkcia je spojitá na intervale, ak jej graf neobsahuje žiadne "prerušenia", "skoky" či "diery" v rámci daného intervalu. Inými slovami, funkcia je spojitá na intervale, keď dokážeme jej graf nakresliť bez zdvihnutia cerzuky z papiera (teda akoby jedným ťahom). 


* Funkcia $f$ je spojitá v bode $x_0$ vtedy, keď $x_1$ je blízko $x_0$, tak aj $f(x_1)$ je blízko $f(x_0)$. Inými slovami blízke body (hodnoty) $x_1$ a $x_0$ sa zobrazia pomocou funkcie $f$ blízko seba tzn. $f(x_1)$ je blízko $f(x_0)$. Ešte ináč povedané, malé zmeny (chyby resp. odchýlky) v nezávislej premennej $x$ vedú k malým zmenám (chybám resp. odchýlkam) funkčných hodnôt $f(x)$. 

Nech $f(x)=x^2$ a predpokladajme, že vieme o spojitosti funkcie $x^2$. Chceme zistiť čomu je rovné $f(\pi)$. Uvedomme si, že číslo $\pi$ má nekonečne veľa číslic za desatinnou bodkou resp. čiakou, a teda nie je možné ho celé dosadiť do funkcie, tzn. nevieme úplne presne určiť $f(\pi)$. Vďaka spojitosti ale máme zaručené, že ak vezmeme napr. číslo $3.14$, ktoré je blízko čisla $\pi$, tak potom $f(3.14)$ bude hodnota blízka hodnote $f(\pi)$, preto stačí vypočítať $f(3.14)$, čo nie je problém. Ak by sme chceli určiť hodnotu $f(\pi)$ presnejšie, tak stačí dosadiť do $f(x)$ miesto $x$ číslo bližšie k $\pi$, než je číslo $3.14$. Takto vieme určiť hodnotu $f(\pi)$ s ľubovoľnou (vopred predpísanou) presnosťou resp. chybou.

#### <font color=blue> Príklad 1 </font>  

```python
x = var('x') 
f1 = 3*x^3-x^2+2*x
f2 = 1/x 
kwds = {'fontsize':10} 
p1 = plot(f1(x), (x, -2, 2), axes_labels=['$x$', '$y$'], axes_labels_size = 1) + text(r"$f_1(x)=3x^3-x^2+2x$", (6, 28), color='blue',**kwds) 
p2 = plot(f2(x), (x, -2, -0.1), axes_labels=['$x$', '$y$'], axes_labels_size = 1) + plot(f2(x), (x, 0.1, 2), axes_labels=['$x$', '$y$'], axes_labels_size = 1) + text(r"$f_2(x)=\dfrac{1}{x}$", (1.5, 10), color='blue',**kwds) 
p3 = line([(0.1,7),(1,7)], color = 'blue', axes_labels=['$x$ (hodiny)', '$y$ (doláre)'], axes_labels_size = 1.0) + line([(1.1,14),(2,14)], color = 'blue') + line([(2.1,21),(3,21)], color = 'blue') + circle((0,7), 0.1, color = 'blue') + circle((1, 7), 0.1, color='blue', fill=True) + circle((1,14), 0.1, color = 'blue') + circle((2, 14), 0.1, color='blue', fill=True) + circle((2,21), 0.1, color = 'blue') + circle((3, 21), 0.1, color='blue', fill=True) + text(r"$p(x)$", (3, 22), color='blue',**kwds) 
G = graphics_array([p1,p2,p3])
# G.show(figsize=[10, 4])  
G.show(figsize=[12, 5], aspect_ratio=0.5) 
```

Pozrime sa na príklady funkcií $f_1$, $f_2$, $p$, ktorých grafy máme vykreslené. Funkcia $f_1$ je spojitá na ľubovoľnom intervale. Funkcia $f_2$ nie je definovaná v bode $0$. Je spojitá na takom intervale, ktorý neobsahuje $0$. Funkcia $p$ vyjadruje nasledovné: Spoločnosť prenajíma autá za 7 dolárov na hodinu ale aj prenájom na menej ako hodinu stojí 7 dolárov (vždy sa platí za celú hodinu); 14 dolárov sa platí za prenájom auta na viac ako jednu a menej ako dve hodiny atď. Ak je $p(x)$  cena za výpožicku auta trvajúcu $x$ hodín, potom je jej graf (viď graf $p(x)$ hore) postupnosťou "schodov". Táto funkcia nie je spojitá na žiadnom otvorenom intervale obsahujúcom kladné celé číslo, pretože graf v týchto bodoch preskočí - nastáva tu skok (viď prázdne a plné krúžky v grafe). Všimnime si ešte, že $p(0.99) = p(1) = 7$ dolárov, zatiaľ čo $p(1.01) = 14$ dolárov, pretože keď prekročíme jednu hodinu (čas zapožičania auta), cena vyskočí na 14 dolárov. Takže malý časový rozdiel môže viesť k významnému rozdielu v nákladoch. Preto $p$ nie je spojitá v $x = 1$. Ako vidíme z grafu, znamená to tiež, že nie je spojitá v žiadnom otvorenom intervale obsahujúcom bod $x = 1$.

#### <font color=blue> Príklad 2 </font> 

Pri štúdiu limít sme nadobudli intuíciu, že **limita popisuje správanie sa funkcie v okolí (v blízkosti) nejakého bodu tzn. limita udáva, k čomu sa blíži funkcia (jej funkčné hodnoty), keď sa blížime ku konkrétnemu bodu po osi $x$**. Napr. $\lim\limits_{x\to1} g(x)=3$ nám hovorí, že keď $x\to1$ tak $g(x)\to3$ resp. keď je $x$ blízko k $1$, tak $g(x)$ je blízko k $3$. Avšak, ako sme už videli v predošlej téme, **limita nemusí dobre vystihovať, čomu je rovná samotná funkčná hodnota**  $g(1)$. Mohlo by sa totiž stať, že napr. funkcia $g(x)$ sa v okolí bodu $1$ blíži k hodnote $3$ ale samotné $g(1)$ je ďaleko od hodnoty $3$ (nastáva "skok" v grafe funkcie), pozrime sa na nasledujuci graf - funkcia je konštantná s hodnotou $3$,  s výnimkou bodu $1$, kde nadobúda funkcia hodnotu $4$ (je tu tzv. "skok"). 

```python
x = var('x') 
kwds = {'fontsize':12} 
p = plot(3, (x, -4, 0.95), axes_labels=['$x$', '$y$'], axes_labels_size = 1, ymin = 0, ymax = 4) + plot(3, (x, 1.05, 4), axes_labels=['$x$', '$y$'], axes_labels_size = 1, ymin = 0, ymax = 4.2) + circle((1, 3), 0.05, color='blue') + circle((1, 4), 0.05, color='blue', fill=True) + text(r"$g(x)$", (2, 3.5), color='blue',**kwds) 
show(p)
```

Ináč povedané, **ak funkcia nie je spojitá, potom malej zmene premennej $x$ odpovedá veľká zmena funkčnej hodnoty $g(x)$**. Čiže ak je funkcia nespojitá v bode $x_0$ a vezme $x_1$ blízke $x_0$, tak $g(x_1)$ bude ďaleko od $g(x_0)$. Keď sa pozrieme konkrétne na graf funkcie $g$ a položíme $x_0=1$, $x_1=0.99$, tak máme $f(x_0)=f(1)=4$ a $f(x_1)=f(0.99)=3$. V tejto časti sa budeme zaoberať takými funkciami, u ktorých toto "nepekné" správanie nenastáva vďaka spojitosti.

#### <font color=blue> Príklad 3 </font> 

Kvôli názornosti a lepšiemu uchopeniu pojmu spojitosti funkcie sa ešte pozrime na nasledujúce tri príklady funkcií $h_1$, $h_2$, $h_3$. Tieto **tri funkcie** majú okrem iného spoločné aj to, že **majú rovnakú limitu v bode** $x_0=1$. **Avšak funkcia $h_1$ v bode $x_0=1$ definovaná nie je**, čo spôsobuje "dieru" v jej grafe práve v tomto bode. **O takej funkcii prehlásime, že nie je spojitá v bode $1$**. Pri funkcii $h_2$ môžeme pozorovať, že zatiaľ čo $\lim\limits_{x\to1}h_2(x)=3$, tak $h_2(1)=2$, a teda **limita v bode $1$ sa nerovná funkčnej hodnote v bode $1$**. **Aj v tomto prípade povieme, že funkcia nie je spojitá v bode $x_0=1$**. Na druhej strane funkcia $h_3$ sa zdá, akoby mala najlepšie vlastnosti, lebo jej **limita a funkčná hodnota v bode $x_0=1$ sa rovnajú**, t.j. $\lim\limits_{x\to1}h_3(x)=3=h_3(1)$. V grafe funkcie $h_3$ nie je žiadna "diera" ani "skok" v bode $x_0=1$. **O takej funkcii povieme, že je spojitá v bode $1$.** 

```python
x = var('x') 
h = -(x-1)^2+3 
kwds = {'fontsize':12} 
p1 = plot(h, (x, -0.5, 0.95), axes_labels=['$x$', '$y$'], axes_labels_size = 1) + plot(h, (x, 1.05, 3), axes_labels=['$x$', '$y$'], axes_labels_size = 1) + circle((1,3), 0.05, color = 'blue') + text(r"$h_1$", (1.5, 3.2), color='blue',**kwds) 
p2 = plot(h, (x, -0.5, 0.95), axes_labels=['$x$', '$y$'], axes_labels_size = 1) + plot(h(x), (x, 1.05, 3), axes_labels=['$x$', '$y$'], axes_labels_size = 1) + circle((1,3), 0.05, color = 'blue') + circle((1, 2), 0.05, color='blue', fill=True) + text(r"$h_2$", (1.5, 3.2), color='blue',**kwds) 
p3 = plot(h, (x, -0.5, 0.95), axes_labels=['$x$', '$y$'], axes_labels_size = 1) + plot(h(x), (x, 1.05, 3), axes_labels=['$x$', '$y$'], axes_labels_size = 1) + circle((1, 3), 0.05, color='blue', fill=True) + text(r"$h_3$", (1.5, 3.2), color='blue',**kwds) 
G = graphics_array([p1,p2,p3])
G.show(figsize=[8, 3])  
# G.show(figsize=[12, 5], aspect_ratio=0.5) 
# show(G)
```

### <font color=red> Definícia 1 </font> 

Nech funkcia $f:\ \mathbb{R} \to \mathbb{R}$ je definovaná na nejakom okolí bodu $x_0$ (aj v bode $x_0$). Hovoríme, že $f$ je **_spojitá_** v bode $x_0$, ak $\lim\limits_{x \to x_0} f(x) = f(x_0)$.

Funkciu $f:\ \mathbb{R} \to \mathbb{R}$ definovanú na nejakom intervale $\langle x_0, x_0+\delta)$  nazveme **_spojitou sprava_** v bode $x_0$ ak $\lim\limits_{x \to x_0^+} f(x) = f(x_0)$. 

Funkciu $f:\ \mathbb{R} \to \mathbb{R}$ definovanú na nejakom intervale $(x_0-\delta, x_0\rangle$  nazveme **_spojitou zľava_** v bode $x_0$ ak $\lim\limits_{x \to x_0^-} f(x) = f(x_0)$. 

--- 

Uvádzame príklad funkcie spojitej sprava a súčasne nespojitej zľava v bode $x_0=1$ (graf vľavo) a tiež príklad funkcie spojitej zľava a súčasne nespojitej sprava v bode $x_0=1$ (graf vpravo). 

```python
x = var('x') 
kwds = {'fontsize':12} 
p1 = plot(sqrt(x), (x, 0, 0.95), axes_labels=['$x$', '$y$'], axes_labels_size = 1) + plot(x^2+1, (x, 1.02, 2), axes_labels=['$x$', '$y$'], axes_labels_size = 1) + circle((1,1), 0.05, color = 'blue') + circle((1,2), 0.05, color = 'blue', fill=True) 
p2 = plot(x^2, (x, 0, 0.98), axes_labels=['$x$', '$y$'], axes_labels_size = 1) + plot(sqrt(x)+1, (x, 1.05, 2), axes_labels=['$x$', '$y$'], axes_labels_size = 1) + circle((1,1), 0.05, color = 'blue', fill=True) + circle((1, 2), 0.05, color='blue') 
G = graphics_array([p1,p2])
G.show(figsize=[10, 4]) 
```

#### Poznámka 1 

* Ak je funkcia $f$ spojitá v bode $x_0$, potom platí: 
    1. $x_0\in\mathcal{D}(f)$; 
    2. funkcia je definovaná na nejakom okolí (**nie prstencovom! rozdiel oproti limite!**) bodu $x_0$, t.j. existuje číslo $\delta>0$ také, že $(x_0-\delta,x_0+\delta)\subset\mathcal{D}(f)$; 
    3. existuje vlastná limita $\lim\limits_{x\to x_0} f(x)=A$; 
    4. $A=f(x_0)$. 
    
* Bod, v ktorom funkcia $f$ nie je spojitá (neplatí aspoň jeden zo štyroch bodov uvedených vyššie) sa nazýva **_bod nespojitosti_** funkcie $f$. 

* **Spojitosť funkcie $f$ v bode $x_0$, definovanej na nejakom okolí bodu $x_0$, je možné vyjadriť formálne (podobne ako limitu) aj takto:** 

$$\left(\forall \varepsilon > 0\right)\left(\exists \delta > 0\right)\left(\forall x \in (x_0 - \delta,x_0+\delta)\right)\ |f(x)-A| < \varepsilon$$ 

    resp. 
    
$$\left(\forall \varepsilon > 0\right)\left(\exists \delta > 0\right)\left(\forall x\in\mathcal{D}(f):|x-x_0| <\delta\right)\ |f(x)-A| < \varepsilon.$$


* **Všimnime si rozdiel!** V definícii limity v bode $x_0$ nehrali samotné $x_0$ a $f(x_0)$ dôležitú úlohu, podstatné bolo správanie sa funkcie na prstencovom okolí bodu $x_0$ (bez $x_0$) a funkcia v bode $x_0$ nemusela byť definovaná, stačilo, že bola definovaná na prstencovom okolí (s možnou výnimkou bodu $x_0$). Na druhej strane, pri definovaní spojitosti v bode $x_0$ sa vyžaduje, aby funkcia bola definovaná aj v samotnom bode $x_0$ (resp. na jeho okolí, už nestačí len prstencové!), a teda $x_0$ aj $f(x_0)$ tu hrajú dôležitú úlohu. 


* Limitu v bode $x_0$ spojitej funkcie $f$ určíme jednoducho ako $f(x_0)$ (Napr. Veta 3 o výpočte limity polynomickej funkcie v téme 1). 


* Väčšina (elementárnych) funkcií, s ktorými sme sa už stretli, sú spojité všade tam, kde sú definované. Napríklad exponenciálne funkcie, polynómy, sínus a kosínus sú všade spojité. Racionálne funkcie sú spojité na akomkoľvek intervale, v ktorom ich menovatele nie sú nulové. Funkcie vytvorené sčítaním, násobením alebo zložením spojitých funkcií sú tiež spojité (viď ďalej Veta 2).

### <font color=green> Veta 2 </font> 

Ak funkcie $f,g$ sá spojité (spojité sprava resp. zľava) v bode $x_0$ a $c \in \mathbb{R}$, tak sú v bode $x_0$ spojité (spojité sprava resp. zľava) aj funkcie $|f|,\ c\cdot f,\ f \pm g,\ f \cdot g$. Ak $g(x_0) \neq 0$, tak v bode  $x_0$ je spojitá (spojitá sprava resp. zľava) aj funkcia $\dfrac{f}{g}$.

#### Poznámka 2

* V prípade zloženia spojitých funkcií je možné rátať limitu celkom pohodlne a to tak, že limitou môžeme "vojsť" do funkcie $f$, ktorá je spojitá v bode $z$, pričom $z=\lim\limits_{x\to x_0}g(x)$, t.j. 
<br/>
$$\lim\limits_{x\to x_0}f\left(g(x)\right)=f\left(\lim\limits_{x\to x_0}g(x)\right).$$


* Presnejšie teda platí tvrdenie, že ak existuje $\lim\limits_{x\to x_0}g(x)=z$ a funkcia $f$ je spojitá v bode $z$, potom v bode $x_0$ existuje limita zloženej funkcie $f\left(g(x)\right)$ a platí $\lim\limits_{x\to x_0}f\left(g(x)\right)=f(z)$. 


* Príklad: 
$$\lim\limits_{x\to 2} e^{-x^2}= e^{\lim\limits_{x\to 2}-x^2}=e^{-2^2}=e^{-4}=\frac{1}{e^4}\approx0.0183,$$
pričom $f\left(g(x)\right)=e^{-x^2}$, $g(x)=-x^2$, $z=\lim\limits_{x\to 2}g(x)=\lim\limits_{x\to 2}(-x^2)=-4$ a funkcia $f(x)=e^{-x}$ je spojitá v bode $z=-4$.

### <font color=green> Veta 5 </font> 

Nech funkcia $f$ je spojitá v bode $x_0$ a $f(x_0) \neq 0$. Potom  existuje okolie $U$ bodu $x_0$ také, že pre všetky $x \in U$ platí $\mathrm{sgn}(f(x)) = \mathrm{sgn}(f(x_0))$, kde $\mathrm{sgn}(y)$ je funkcia nadobudajúca hodnoty iba $\{-1,0,1\}$ podľa toho, či je $y<0$, $y=0$ alebo $y>0$.

--- 

Funkcia $\mathrm{sgn}(x)$ vlastne určuje "znamienko" hodnoty $x$ a Veta 5 teda hovorí o tom, že ak je funkcia $f$ spojitá v bode $x_0$, potom existuje okolie bodu $x_0$ také, že funkcia $f$ na ňom nemení znamienko (všetky funkčné hodnoty $f(x)$ na tomto okolí sú buď kladné alebo záporné a ich znamienko je totožné so znamienkom hodnoty $f(x_0)$).

#### <font color=blue> Príklad 5 </font> 

Funkcia $f(x)=\cos x$ je spojitá na $\mathbb{R}$ a teda aj v bode $x_0=0$. Zrejme $\cos\,(0)=1\neq0$, preto podľa Vety 5 existuje okolie $U$ bodu $x_0=0$ také, že pre všetky $x$ z tohto okolia sú znamienka hodnôt $f(x)$ rovnaké ako znamienko hodnoty $f(x_0)=f(0)$. Nasledujúci príklad nám ilustruje situáciu, keď **z nespojitej funkcie môžeme "vyrobiť" spojitú funkciu**. Ide o to, že ak existuje vlastná limita $\lim\limits_{x\to x_0} f(x)=A$ ale bod $x_0$ nepatrí do definičného oboru funkcie $f$, potom môžeme funkciu $f$ dodefinovať v bode $x_0$ tak, aby bola spojitá v tomto bode. Z (nespojitej) funkcie $f$ vytvoríme novú (spojitú) funkciu $\tilde{f}$ takto: 

$$\tilde{f}=\begin{cases}
f(x) & \quad \text{pre } x\neq x_0, \\ 
A & \quad \text{pre } x=x_0.
\end{cases}$$

### <font color=green> Veta 7 (Weierstrassova o maxime a minime) </font> 

Nech funkcia $f$ je definovaná a spojitá na uzavretom intervale $\langle a,b \rangle$. Potom $f$ nadobúda na $\langle a,b \rangle $ maximum a minimum, t.j. 

$$(\exists c,d \in \langle a,b \rangle)(\forall x \in \langle a,b\rangle )\ f(c) \leq f(x) \leq f(d).$$

Ďalšia veta (Bolzanova) pojednáva o tom, že ak máme spojitú funkciu $f$ na uzavretom intervale $\langle a,b \rangle$ a funkčné hodnoty v krajných bodoch intervalu majú rôzne znamienka t.j. $f(a)\cdot f(b)<0$, potom musí graf funkcie $f$ pretínať os $x$ na intervale $\langle a,b \rangle$. Inými slovami, existuje $c \in \langle a,b \rangle$ také, že platí $f(c)=0$, tzn. $c$ je nulový bod funkcie $f$ na intervale $\langle a,b \rangle$.

### <font color=green> Veta 8 (Bolzano) </font> 

Nech funkcia $f$ je definovaná a spojitá na intervale $\langle a,b \rangle$ a nech $\mbox{sgn}(f(a)) \neq \mbox{sgn}(f(b))$. Potom existuje $ \xi \in ( a,b )$ také, že $ f(\xi) = 0$.

Z grafu tiež vidieť, že $\mathrm{sgn}(f(-\pi/2))\neq\mathrm{sgn}(f(\pi/2))$ lebo $f(-\pi/2)<0$ a $f(\pi/2)>0$. Podľa Bolzanovej vety existuje $\xi \in \left(-\pi/2,\pi/2\right)$ také, že $f(\xi)=0$. V našom prípade je $\xi=0$.

Bolzanova veta je základom tzv.  **_metódy polenia intervalov_** (metóda bisekcie), ktorá sa využíva pre numerické (približné) riešenie rovnice $f(x) = 0$ s  požadovanou presnosťou (resp. chybou resp. toleranciou). 

**VSTUP:** 
- funkcia $f$ spojitá na množine $J$; 
- interval $I = \langle a,b \rangle \subseteq J$ taký, že $\mathrm{sgn}(f(a)) \neq \mathrm{sgn}(f(b))$;  
- presnosť $\varepsilon$. 
		

**VÝSTUP:** 
- približné riešenie $x^*$ (s presnosťou $\varepsilon$) rovnice $f(x) = 0$ 
		
**POSTUP:**
- $k:=0,\,I_k:= I,\, a_k := a,\, b_k := b$; 
- pokiaľ $|a_k - b_k| > 2\varepsilon$, opakuj: 
    - $c_k:=\dfrac{a_k + b_k}{2}$; 
    - ak $\mathrm{sgn}(f(a_k)) \neq \mathrm{sgn}(f(c_k))$ tak $a_{k+1}:=a_k, b_{k+1}:= c_k$ inak $a_{k+1} := c_k,b_{k+1}:=b_k$; 
    - $k:= k+1$; 
    
- $x^* := c_k$.

<ins>Výstupy</ins>: eps - presnosť;  $c$ - približná hodnota koreňa; $f(c)$ - funkčná hodnota v nájdenom približnom riešení rovnice (mala by byť blízka nule); iterations - počet krokov (iterácií) potrebných k nájdeniu približného riešenia s danou presnosťou; grafický výstup v podobe znázornených intervalov nájdených v jednotlivých krokoch riešenia. 


```python
# created by by William Stein 

def bisect_method(f, a, b, eps):
    try:
        f = f._fast_float_(f.variables()[0])
    except AttributeError:
        pass
    intervals = [(a,b)]
    two = float(2); eps = float(eps)
    while True:
        c = (a+b)/two
        fa = f(a); fb = f(b); fc = f(c)
        if abs(fc) < eps: return c, intervals
        if fa*fc < 0:
            a, b = a, c
        elif fc*fb < 0:
            a, b = c, b
        else:
            raise ValueError("f must have a sign change in the interval (%s,%s)"%(a,b))
        intervals.append((a,b))
pretty_print(html("<h1>Double Precision Root Finding Using Bisection</h1>"))
@interact
def _(f = cos(x) - x, a = float(0), b = float(1), eps=(-3,(-16, -1))):
     eps = 10^eps
     print("eps = %s" % float(eps))
     try:
         c, intervals = bisect_method(f, a, b, eps)
     except ValueError:
         print("f must have opposite sign at the endpoints of the interval")
         show(plot(f, a, b, color='red'), xmin=a, xmax=b)
     else:
         print("c =", c)
         print("f(c) = %r" % f(x=c))
         print("iterations =", len(intervals))
         P = plot(f, a, b, color='red')
         h = (P.ymax() - P.ymin())/ (1.5*len(intervals))
         L = sum(line([(c,h*i), (d,h*i)]) for i, (c,d) in enumerate(intervals) )
         L += sum(line([(c,h*i-h/4), (c,h*i+h/4)]) for i, (c,d) in enumerate(intervals) )
         L += sum(line([(d,h*i-h/4), (d,h*i+h/4)]) for i, (c,d) in enumerate(intervals) )
         show(P + L, xmin=a, xmax=b)
```

Veta 9 hovorí, že funkcia, ktorá je spojitá na intervale $\langle a,b \rangle$ musí nadobúdať všetky hodnoty, ktoré sú medzi hodnotami $f(a)$ a $f(b)$.

### <font color=green> Veta 9 (Darboux) </font> 

Nech funkcia $f$ je definovaná a spojitá na intervale $\langle a,b \rangle$ a nech $f(a) \neq f(b)$. Nech $d$ je ľubovoľné číslo také, že $f(a) < d < f(b)$ resp. $f(b) < d < f(a)$. Potom existuje $ c \in (a,b))$ také, že $f(c) = d$.

### <font color=green> Veta 10 </font> 

Nech funkcia $f$ je definovaná a spojitá na intervale (ľubovoľného typu) $I$. Potom obraz intervalu $I$ pri funkcii $f$ je buď jednobodová množina alebo interval.

#### <font color=blue> Príklad 11 </font> 

Uvažujme napr. funkciu $f(x)=\sin x$ na intervale $\langle 0, 2\pi )$ (vykreslite si graf). Na tomto intervale je uvažovaná funkcia spojitá a nadobúda všetky hodnoty od $-1$ po $1$, t.j. $\langle -1, 1 \rangle$. Symbolicky by sme to mohli vyjadriť takto: "$f\left(\langle 0, 2\pi )\right)=\langle -1, 1 \rangle$", to značí, že obrazom intervalu $\langle 0, 2\pi )$ pri funkcii $f$ je interval $\langle -1, 1 \rangle$. 

---

### <font color=green> Veta 11 </font> 

Ak $f$ je rýdzomonotónna a spojitá funkcia na intervale $I$ a $J$ je obraz intervalu $I$ pri funkcii $f$ (t.j. "$J,=f(I)$"), tak inverzná funkcia $f^{-1}$ je spojitá na intervale $J$.

---

---

#### <font color=blue> Príklad 12 </font> 

Uvažujme napr. funkciu $f(x)=\sin x$ na intervale $\langle -\pi/2, \pi/2 )$ (vykreslite si graf). Na tomto intervale je uvažovaná funkcia rýdzomonotónna a spojitá. Obrazom intervalu $I=\langle -\pi/2, \pi/2 )$ pri funkcii $f$ je interval $J=\langle -1, 1 \rangle$. Na $I=\langle -\pi/2, \pi/2 )$ existuje k $f(x)=\sin x$ inverzná funkcia $f^{-1}(x)=\arcsin x$ (vykreslite graf). Funkcia $f^{-1}(x)=\arcsin x$ je podľa Vety 11 spojitá na $J=\langle -1, 1 \rangle$.

### <font color=green> Veta 12 </font> 

Nech funkcia $f$ je definovaná a monotónna na intervale $I$ a nech obor hodnôt funkcie $f$ na intervale $I$ je jednoprvková množina alebo interval. Potom funkcia $f$ je spojitá na intervale $I$.
