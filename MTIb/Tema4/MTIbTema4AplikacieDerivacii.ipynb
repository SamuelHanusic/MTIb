 # <font color=brown> Výpočet špecifických limít funkcií </font> 

Ak existuje $\lim\limits_{x\to x_0}f(x)$, $\lim\limits_{x\to x_0}g(x)$ a $\lim\limits_{x\to 0}g(x)\ne0$, tak existuje aj limita podielu  $\dfrac{f(x)}{g(x)}$ rovná podielu obidvoch limít (téma 1, Veta 4, časť e) ). V prípade, že $\lim\limits_{x\to x_0}f(x)$ existuje a $\lim\limits_{x\to x_0}g(x) = \pm \infty$, limita podielu $\dfrac{f(x)}{g(x)}$ je rovná $0$; v prípade, že $\lim\limits_{x\to x_0}f(x) = \pm \infty$ a $\lim\limits_{x\to x_0}g(x)$ existuje a je nenulová, limita podielu je (až na znamienko) $\infty$ resp. neexistuje (v zmysle, že existujú len navzájom rôzne nevlastné jednostranné limity).

Prípady, kedy o limite podielu $\dfrac{f(x)}{g(x)}$ nemožno rozhodnúť podľa uvedených tvrdení, sú prípady, keď $\lim\limits_{x \to x_0} f(x) = \lim\limits_{x \to x_0} g(x) = 0$ alebo keď $\lim\limits_{x \to x_0} f(x) = \lim\limits_{x \to x_0} g(x) = \pm \infty$ (tzv. limity typu "$\frac{0}{0}$" alebo typu "$\frac{\infty}{\infty}$"). Uvedené prípady rieši za určitých predpokladov nasledujúca veta. 

---

### <font color=green> Veta 1 (L'Hospitalovo pravidlo) </font> 

Nech $f$, $g$ sú funkcie, $\lim\limits_{x\to x_0}f(x)=\lim\limits_{x\to x_0}g(x)=0$ alebo $\lim\limits_{x\to x_0}f(x)=\lim\limits_{x\to x_0}g(x)=\pm \infty$. Nech existuje $\lim\limits_{x\to x_0}\dfrac{f'(x)}{g'(x)}$ (vlastná alebo nevlastná). Potom existuje aj $\lim\limits_{x\to x_0}\dfrac{f(x)}{g(x)}$ a tieto dve limity sú rovnaké. 

--- 

--- 

#### Poznámka 1 

* Ak limita $\lim\limits_{x\to x_0}\dfrac{f'(x)}{g'(x)}$ neexistuje, tak o limite $\lim\limits_{x\to x_0}\dfrac{f(x)}{g(x)}$ nemožno nič tvrdiť! (na základe Vety 1) 


* Pri L'Hospitalovom pravidle  výraz za znakom limity **nederivujeme ako podiel**, ale čitateľa a menovateľa zvlášť! 


* L'Hospitalovo pravidlo je možné aplikovať aj v prípade jednostranných limít. 

---

---

#### <font color=blue> Príklad 1 </font> 

Vypočítajte $\lim\limits_{x\to 0}\dfrac{1-\cos x}{x^2}$.  

<ins> Riešenie: </ins> 

Keďže $\lim\limits_{x\to0}(1-\cos x)=0$, $\lim\limits_{x\to0}x^2=0$, môžeme použiť L'Hospitalovo pravidlo. Dostávame tak 

$$\lim\limits_{x\to 0}\dfrac{1-\cos x}{x^2} = \lim\limits_{x\to0}\dfrac{(1-\cos x)'}{(x^2)'}=\lim\limits_{x\to0}\dfrac{\sin x}{2x}=\frac12\lim\limits_{x\to0}\dfrac{\sin x}x=\frac12\cdot1=\frac12.$$ 

--- 

---

#### <font color=blue> Príklad 2 </font> 

Vypočítajte $\lim\limits_{x\to\infty}\dfrac x{e^x}$. 

<ins> Riešenie: </ins> 

Keďže $\lim\limits_{x\to \infty}x=\infty$, $\lim\limits_{x\to\infty}e^x = \infty$, môžeme použiť L'Hospitalovo pravidlo. Dostávame tak 

$$\lim\limits_{x\to\infty}\dfrac x{e^x}=\lim\limits_{x\to\infty}\dfrac{(x)'}{(e^x)'}=\lim\limits_{x\to\infty}\dfrac1{e^x}=0.$$ 

--- 

--- 

#### Poznámka 2 

V niektorých prípadoch L'Hospitalovo pravidlo nemusí viesť k výsledku: jeho opakovaným použitím pri výpočte $\lim\limits_{x\to\infty}\dfrac{\sqrt{x^2-1}}x$ dostávame

$$
	\lim_{x\to\infty}\frac{\sqrt{x^2-1}}x = \lim_{x\to\infty}\frac{\frac{2x}{2\sqrt{x^2-1}}}{1} = \lim_{x\to\infty}\frac {x}{\sqrt{x^2-1}}=$$ $$ \lim_{x\to\infty}\frac{1}{\frac{2x}{2\sqrt{x^2-1}}} =\lim_{x\to\infty}\frac{\sqrt{x^2-1}}x=\dots\,.
$$

Opakovaným použitím L'Hospitalovo pravidla dostávame rovnakú limitu ako na začiatku a výpočet sa "zacyklí".

Ľahko ale vidieť, že
$$
	\lim_{x\to\infty}\frac{\sqrt{x^2-1}}x= \lim\limits_{x \to \infty} \sqrt{\frac{x^2 - 1}{x^2}} = \lim\limits_{x \to \infty} \sqrt{\frac{x^2}{x^2}-\frac{1}{x^2}}  = \lim_{x\to\infty}\sqrt{1-\frac1{x^2}}=\sqrt1=1\,.
$$

---

<font color=orange> Prečo platí $\lim_{x\to\infty}\sqrt{1-\frac1{x^2}}=\sqrt1$ ? Je to časť výpočtu v Poznámke 2. Pomôžte si tvrdeniami o limitách z predchádzajúcich tém. </font> 

**L'Hospitalovo pravidlo možno niekedy využiť aj pri výpočte limít, ktoré nie sú typu "$\frac{0}{0}$" resp. "$\frac{\infty}{\infty}$":**

* $\lim\limits_{x \to x_0} f(x)\cdot g(x)$ je typu "$0\cdot\infty$", t.j. $\lim\limits_{x \to x_0} f(x) = 0$ a $ \lim\limits_{x \to x_0} g(x) = \pm\infty$, tak možno využiť úpravu
	
    $$
	f(x)\cdot g(x) = \dfrac{f(x)}{\frac{1}{g(x)}} = \dfrac{g(x)}{\frac{1}{f(x)}}
	$$
	
    čím sa výpočet transformuje na výpočet limity typu "$\frac{0}{0}$" alebo typu "$\frac{\infty}{\infty}$". 
    
    
* ak $\lim\limits_{x \to x_0} (f(x) - g(x))$ je typu"$\infty - \infty$", t.j. $\lim\limits_{x \to x_0} f(x) =$ $ \lim\limits_{x \to x_0} g(x) = +\infty$, tak možno využiť úpravu
	
    $$
	f(x) - g(x) = \dfrac{\frac{1}{g(x)} - \frac{1}{f(x)}}{\frac{1}{f(x)g(x)}}
	$$
	
    čím sa výpočet transformuje na výpočet limity typu "$\frac{0}{0}$".


* ak $\lim\limits_{x \to x_0} f(x)^{g(x)}$ je typu "$0^0$", "$\infty^0$" alebo "$1^\infty$", tak možno využiť úpravu
	
    $$
	f(x)^{g(x)} = e^{\ln(f(x)^{g(x)})} = e^{g(x)\ln f(x)}
	$$
	
    $$ \lim\limits_{x \to x_0} f(x)^{g(x)} = e^{\lim\limits_{x \to x_0}g(x)\ln f(x)} 
	$$
	
    čím sa výpočet transformuje na výpočet limity typu "$0\cdot \infty$" v exponente a tu uplatníme prvý bod tohto odseku.

Vypočítajte limitu $\lim\limits_{x\to0^+}x^x$.

<ins> Riešenie: </ins> 

Upravíme výraz $x^x$ (využitím vzťahu $x^x = e^{\ln\left ( x^x \right )}$) a použijeme L'Hospitalovo pravidlo: 

$$\lim\limits_{x\to0^+}x^x=\lim\limits_{x\to0^+}e^{x\ln x}=e^{\lim\limits_{x\to0^+}x\ln x}=e^{\lim\limits_{x\to0^+}\frac{\ln x}{\frac1x}}=e^{\lim\limits_{x\to0^+}\frac{\frac1x}{-\frac1{x^2}}}=e^{\lim\limits_{x\to0^+}(-x)}=e^0=1.$$

 # <font color=brown> Aproximácia hodnôt funkcie </font> 

Spomeňme si, že derivácia funkcie $f$ môže poslúžiť na nájdenie smernice dotyčnice ku grafu funkcie $f$. V bode $x=x_0$ má dotyčnica $l$ rovnicu $l:\ y=f(x_0)+f'(x_0)\cdot(x-x_0)$. Pomocou dotyčnice je možné aproximovať hodnoty $f(x)$ pre $x$ blízko $x_0$. 

Predpokladajme, že chceme pomocou (hodnoty) dotyčnice aproximovať (približne vyčísliť) hodnotu $\sin\,(1.1)$. Potrebujeme k tomu dotyčnicu k funkcii $f(x)=\sin\,x$ napr. v bode $x_0=\frac{\pi}{3}\approx 1.05$. Podstatné je aby bod, v ktorom dotyčnicu zostrojíme bol blízko bodu $1.1$. Zrejme $1.05$ je blízko k $1.1$. Vieme, že $\sin\,(\pi/3)=\sqrt{3}/2\approx 0.866$ a $\left(\sin\,x\right)'=\cos\,x$, pričom $\cos\,(\pi/3)=1/2$. Preto dotyčnica k $f(x)=\sin\,x$ v bode $x_0=\pi/3$ je 

$$l:y=\ \frac{1}{2}\left(x-\frac{\pi}{3}\right)+0.866 \qquad \text{ resp. } \qquad l(x)=\frac{1}{2}\left(x-\frac{\pi}{3}\right)+0.866.$$ 


$$\text{Zdroj: Hartman (2018).}$$ 

Na obrázku (a) je znázornený graf funkcie $f(x)=\sin\,x$ (modrá farba) spolu s dotyčnicou v bode $x_0=\pi/3$ (červená farba). Malý obdĺžnik znázorňuje oblasť, ktorá je zväčšená (priblížená) na obrázku (b). Práve na obrázku (b) môžeme vidieť ako je hodnota $\sin\,(1.1)$ aproximovaná pomocou hodnoty dotyčnice vyčíslenej v bode $1.1$ t.j. $l(1.1)$. Obidva obrázky ukazujú, ako blízke sú tieto hodnoty resp. aká dobrá je to aproximácia (červená priamka je blízko modrej krivky v bode $1.1$). Aproximácia hodnoty $\sin\,(1.1)$ pomocou hodnoty dotyčnice $l(1.1)$ vyzerá nasledovne: 

$$l(1.1)=\frac{1}{2}(1.1-\pi/3)+0.866=\frac{1}{2}(0.053)+0.866=0.8925.$$ 

Len pre porovnanie $\sin\,(1.1)=0.8912\ldots$ . 

Teraz sa pokúsime ilustrovaný koncept zovšeobecniť. Pre danú funkciu $f(x)$ a bod $x_0$ je dotyčnica k $f$ v bode $x_0$ daná predpisom $l(x)=f'(x_0)(x-x_0)+f(x_0)$. Zrejme $f(x_0)=l(x_0)$. Nech $\Delta x$ je malé číslo vyjadrujúce malú zmenu v hodnote $x$, napr. nech $\Delta x = x - x_0$ resp. $x=x_0+\Delta$. Keďže $x=x_0+\Delta$ je blízko k $x_0$ a dotyčnica v $x_0$ dobre aproximuje hodnoty $f(x)$ pre $x$ blízke $x_0$, tak môžeme tvrdiť že $f(x_0+\Delta x)$ je porovnateľné (blízke) s $l(x_0+\Delta x)$, čo označíme 

$$f(x_0+\Delta x) \approx l(x_0+\Delta x).$$ 

Keď sa hodnota $x_0$ zmení na $x_0+\Delta x$, potom sa hodnota $f$ zmení z $f(x_0)$ na $f(x_0+\Delta x)$ t.j. hodnota $f$ sa zmení o $\Delta f(x_0)=f(x_0+\Delta x)-f(x_0)$. Ak nahradíme $f(x_0+\Delta x)$ aproximáciou pomocou dotyčnice tak dostávame 

$$\Delta f(x_0) \approx l(x_0+\Delta x)-f(x_0)=f'(x_0)\left((x_0+\Delta x)-x_0\right)+f(x_0)-f(x_0)=f'(x_0)\Delta x.$$ 

Kľúčový je posledný výraz resp. približná rovnosť $\Delta f(x_0) \approx f'(x_0)\Delta x = f'(x_0)(x-x_0)$. V jednoduchosti povedané nám to hovorí, že ak sa zmení $x_0$ na $x_0+\Delta x$, potom hodnota funkcie $f$ sa zmení približne o hodnotu $f'(x_0)(x-x_0)$. Výraz $f'(x_0)(x-x_0)$ je natoľko dôležitý, že si zaslúži špeciálne pomenovanie a hovorí o tom nasledujúca definícia. 

---

### <font color=red> Definícia 1 </font> 

Nech funkcia $f$ má v bode $x_0$ deriváciu. Potom výraz $\mbox{d}f(x_0,x) = f'(x_0)(x - x_0)$ sa nazýva **_diferenciál_**  funkcie $f$ v bode $x_0$ pre prírastok $x - x_0$. 

--- 

Z geometrického hľadiska, diferenciál funkcie $f$ v bode $x_0$ pre prírastok $x - x_0$ zodpovedá rozdielu funkčnej hodnoty v bode $x$ pre funkciu $y = f'(x_0)(x-x_0) + f(x_0)$ a funkčnej hodnoty $f(x_0)$ (t.j. prírastok lineárnej funkcie vyjadrujúcej rovnicu dotyčnice ku grafu $f$ v bode $[x_0,f(x_0)]$). 


$$\text{Zdroj: Madaras (2019).}$$

Skúmajme vzťah diferenciálu $\mbox{d}f(x_0,x)$ voči funkčným hodnotám $f(x_0),f(x)$.
Ak označíme $\tau(x_0,x) = \Delta f(x_0) - \mbox{d}f(x_0,x) = (f(x) - f(x_0)) - \mbox{d}f(x_0,x)$, tak platí 

$$\lim\limits_{x \to x_0} \tau(x_0,x) = \lim\limits_{x \to x_0} ((f(x) - f(x_0)) - \mbox{d}f(x_0,x)) = 0 - \lim\limits_{x \to x_0} f'(x_0)(x - x_0) = 0-0 = 0.$$ 

Z toho  vyplýva, že **pre $x$ dostatočne blízke $x_0$ je možné funkčnú hodnotu $f(x)$ aproximovať výrazom $f(x_0) + f'(x_0)(x - x_0)$** (pričom relatívna chyba $\dfrac{|\Delta f(x_0) - \mbox{d}f(x_0)|}{|f(x_0)|}=\dfrac{|f(x)-(f(x_0) + f'(x_0)(x - x_0))|}{|f(x_0)|}$ aproximácie skutočnej hodnoty $f(x_0)$ závisí len na rozdieli $x - x_0$).

### <font color=red> Definícia 2 </font>  
 
Funkcia $f$ sa nazýva **_rastúca v bode_** $x_0$, ak existuje $\delta>0$ také, že pre všetky $x \in (x_0 - \delta, x_0)$ platí $f(x) < f(x_0)$ a pre všetky $x \in (x_0,x_0+\delta)$ platí $f(x) > f(x_0)$. 
 
Funkcia $f$ sa nazýva **_klesajúca v bode_** $x_0$, ak existuje $\delta>0$ také, že pre všetky $x \in (x_0 - \delta, x_0)$ platí $f(x) > f(x_0)$ a pre všetky $x \in (x_0,x_0+\delta)$ platí $f(x) < f(x_0)$.

--- 

**Definíciu 2 je možné si predstaviť** (viď nasledujúci graf) **na** jednoduchom **príklade funkcie $f(x)=x\cdot\ln x$**. V Definícii 2 položme $x_0=1$ a overíme, že funkcia $f(x)=x\cdot\ln x$ je rastúca v tomto bode v zmysle Definície 2. Vezmime $\delta=0.2$, čím dostaneme ľavé (prstencové) okolie $(x_0-\delta,x_0)=(0.8,1)$ bodu $x_0=1$ a tiež pravé (prstencové) okolie $(x_0,x_0+\delta)=(1,1.2)$ bodu $x_0=1$. Pozrime sa najprv na zvolené ľavé (prstencové) okolie bodu $x_0=1$. Zrejme $\forall x\in(0.8,1)$ platí, že $f(x)<f(x_0)$, konkrétne napr. pre $x=0.9$ máme $f(x)=f(0.9)=−0.0948<0=f(1)=f(x_0)$, čiže $f(0.9)<f(1)$. Teraz preskúmajme zvolené pravé (prstencové) okolie bodu $x_0=1$. **Čo nám teda derivácia funkcie hovorí z vizuálneho hľadiska?** 

Tam, kde je $f'$ kladná, dotyčnica k $f$ smeruje nahor. Tam kde je $f'$ záporná, dotyčnica k $f$ smeruje nadol. Vidíme teda, že znamienko derivácie $f'$ nám hovorí, či je funkcia $f$ rastúca (vtedy je $f'>0$) alebo klesajúca (vtedy je $f'<0$) alebo konštantná (vtedy $f'=0$).

Ináč povedané, ak $f'$ má veľkú hodnotu (kladnú/zápornú), potom graf funkcie $f$ je strmý (prudko rastie/klesá smerom nahor/nadol). Na druhej strane ak má derivácia $f'$ malú hodnotu tak graf funkcie $f$ je miernejší, nie je veľmi strmý (funkcia $f$ mierne rastie/klesá).  # <font color=brown> Hľadanie extrémov funkcie</font>

V kurze MTIa sme zaviedli pojmy ako lokálne minimum/maximu a globálne minimum/maximum funkcie. Takým bodom hovoríme (lokálne/globálne) extrémy funkcie. Len si pripomeňme, že funkcia $f$ má lokálne minimum v bode $x_0$ svojho definičného oboru ak $f(x_0)$ je menšie alebo rovné ako hodnoty $f(x)$ pre všetky $x$ z nejakého okolia (otvoreného intervalu obsahujúceho $x_0$) bodu $x_0$. Na druhej strane funkcia $f$ má lokálne maximum v bode $x_0$ svojho definičného oboru ak $f(x_0)$ je väčšie alebo rovné ako hodnoty $f(x)$ pre všetky $x$ z nejakého okolia bodu $x_0$. Globálne minimum resp. maximum dostaneme tak, že požiadavku okolia zmeníme, konkrétne aby $f(x_0)\le f(x)$ resp. $f(x_0)\ge f(x)$ pre všetky $x$ z definičného oboru funkcie $f$.  

Keď hľadáme extrémy funkcie, častokrát nás zaujímajú také body funkcie, ako sú vyznačené na nasledujúcom grafe funkcie $f(x)=x^3-9x^2-48x+52$. Tieto dva body majú spoločnú vlastnosť - dotyčnice ku grafu $f$ v týchto bodoch sú rovnobežné s osou $x$, čo znamená, že derivácie v týchto bodoch sú nulové. Také body spolu s bodmi, v ktorých neexistuje derivácia sú dôležité pri hľadaní extrémov funkcie, preto majú aj špecifické pomenovanie - viď Definícia 3. 

```python
# graf + nulova derivacia resp. dotycnica rovnobezna s osou x 
x = var('x') 
f(x) = x^3-9*x^2-48*x+52
kwds = {'fontsize':11} 
plot(f(x), (x, -5, 13), axes_labels=['$x$', '$y$'], axes_labels_size = 1) + plot(f(x), (x, 0.083, 2), axes_labels=['$x$', '$y$'], axes_labels_size = 1) + point((-2, 104), size=20, color='blue') + point((8, -396), size=20, color='blue') + text(r"$f(x)=x^3-9x^2-48x+52$", (8, 100), color='blue',**kwds) + text(r"$-2$", (-2, -20), color='black',**kwds) + text(r"$8$", (8, 20), color='black',**kwds) + line([(8,0),(8,f(8))], linestyle = 'dashed', color = 'black') + line([(0,f(8)),(8,f(8))], linestyle = 'dashed', color = 'black') + line([(-2,0),(-2,f(-2))], linestyle = 'dashed', color = 'black') + line([(0,f(-2)),(-2,f(-2))], linestyle = 'dashed', color = 'black') + line([(-5,f(-2)),(1,f(-2))], linestyle = 'solid', color = 'red') + line([(5,f(8)),(11,f(8))], linestyle = 'solid', color = 'red')
```

### <font color=red> Definícia 4 </font>  
 
Nech funkcia $f$ je definovaná v okolí bodu $x_0$ a má v bode $x_0$ deriváciu. Potom $f$ sa nazýva **_konvexná v bode_** $x_0$, ak existuje také okolie $U$ bodu $x_0$, že dotyčnica v bode $x_0$ je na $U$ **celá** (okrem bodu $[x_0,f(x_0)]$) **pod** grafom $f$. Podobne, $f$ sa nazýva **_konkávna v bode_** $x_0$, ak dotyčnica v bode $x_0$ je na $U$ **celá** (okrem bodu $[x_0,f(x_0)]$) **nad** grafom $f$.

---

$$\text{Zdroj: Madaras (2019).}$$

$$\text{Zdroj: Madaras (2019).}$$

### <font color=red> Definícia 5 </font>  
 
Bod $x_0$ sa nazýva **_inflexný bod_** funkcie $f$, ak $f$ je definovaná v okolí bodu $x_0$, existuje $f'(x_0)$ (vlastná alebo nevlastná) a existuje také okolie $U$ bodu $x_0$, že  pre všetky $x\in U$, $x<x_0$ je funkcia $f$ konvexná (resp. konkávna) a pre všetky $x\in U$, $x>x_0$ je funkcia $f$ konkávna (resp. konvexná). 

--- 

--- 

#### Poznámka 7 

- V definícii inflexného bodu sa vyžaduje, aby v bode $x_0$ existovala dotyčnica; teda aj keď sa v bode $x_0$ mení konvexnosť na konkávnosť (resp. naopak), ale ak v bode $x_0$ má funkcia "hrot" (t.j. nemá v ňom dotyčnicu), tak takýto bod **nepovažujeme** za inflexný bod (viď nasledujúci obrázok).


$$\text{Zdroj: Madaras (2019).}$$ 

- **Upozornenie !** Nie každý bod $x_0$, v ktorom $f''(x_0)=0$ (alebo $f''$ nie je definovaná) je inflexný bod (podobne ako nie v každom bode, kde $f'=0$ je lokálne maximum alebo minimum). Napr. v prípade $f(x)=x^4$ máme $f''(x)=12x^2$, čiže $f''(0)=0$ ale $f''>0$ pre $x<0$ ale tiež pre $x>0$ a teda nedochádza k zmene konvexnosti (konkávnosti) funkcie v bode $0$ (viď graf), preto bod $0$ nie je inflexný bod. Funkcia $f(x)=x^4$ je konvexná na celom definičnom obore. 


```python
x = var('x') 
f(x) = x^4 
kwds = {'fontsize':12} 
plot(f(x), (x, -2, 2), axes_labels=['$x$', '$y$'], axes_labels_size = 1) + point((0, 0), size=40, color='blue') + text(r"$f(x)=x^4$", (1.2, 10), color='blue',**kwds)  
```

Podľa Tvrdenia 2 to znamená, že funkcia $g$ najprv rastie (t.j. $f'>0$) smerom k lokálnemu maximu v bode $1$ s hodnotou $1/e$ a následne klesá (t.j. $f'<0$). Čo sa týka extrémov funkcie, potrebujeme overiť, či extrém nie je v krajnom bode intervalu, čo je v našom prípade bod $0$, keďže uvažujeme $x\ge 0$. ### <font color=red> Definícia 6 </font>  
 
Priamka $x = a$ sa nazýva **_asymptota bez smernice_** (niekedy tiež vertikálna asymptota) grafu funkcie $f$, ak $f$ má v bode $a$ nevlastnú limitu sprava alebo zľava.

--- 

--- 

#### Poznámka 8 

Ak $x=a$ je asymptota bez smernice (prerušovaná čiara resp. priamka na obrázku), tak hodnota $f(a)$ môže alebo nemusí byť definovaná. 


$$\text{Zdroj: Madaras (2019).}$$ 

--- 

---

### <font color=red> Definícia 7 </font>  
 
Nech funkcia $f$ je definovaná na intervale $(a,\infty)$ (resp. $-\infty, a)$) pre nejaké $a \in \mathbb{R}$. Priamka $y = kx + q$ sa nazýva **_asymptota so smernicou_** grafu funkcie $f$ pre $x \to \infty$ (resp. $x \to -\infty$), ak $\lim\limits_{x \to \infty} (f(x) - (kx+q)) = 0 $ (resp. $\lim\limits_{x \to -\infty} (f(x) - (kx+q)) = 0 $). 

--- 

--- 

#### Poznámka 9 

* Ak pre definičný obor $f$ platí ${\cal{D}}(f) \subseteq (-\infty,a)$ pre nejaké $a \in \mathbb{R}$, tak funkcia $f$ nemá asymptotu so smernicou pre smer $\infty$, a podobne, ak ${\cal{D}}(f) \subseteq (a,\infty)$, tak $f$ nemá asymptotu so smernicou pre smer $-\infty$. 


$$\text{Zdroj: Madaras (2019).}$$ 


* Pokiaľ je smernica $k=0$, tak dostávame priamku v tvare $y=q$, čo je vlastne horizontálna asymptota (rovnobežná s osou $x$). 

--- 

Nasledujúce tvrdenie dobre poslúži pri hľadaní asymptoty so smernicou. 

---

#### <font color=green> Tvrdenie 8 </font> 

Priamka $y=kx+q$ je asymptotou so smernicou ku funkcii $f$ pre $x\to\infty$ (pre $x\to-\infty$) práve vtedy, keď
 
$$k=\lim\limits_{x\to\infty}\dfrac{f(x)}x \qquad (k=\lim\limits_{x\to-\infty}\dfrac{f(x)}x)$$

$$q=\lim\limits_{x\to\infty}(f(x)-k\cdot x) \qquad (q=\lim\limits_{x\to-\infty}(f(x)-k\cdot x)).$$ 

--- 

--- 

#### Poznámka 10 

Funkcia môže mať dve **rôzne** asymptoty so smernicou, preto je dôležité vypočítať príslušné limity aj pre $x \to -\infty$ (zvlášť v prípade, keď predpis funkcie obsahuje exponenciálne resp. cyklometrické funkcie).

```python
# menovatel zlomku musi byt rozny od nuly, preto zistime, kedy sa menovatel rovna nule t.j. vyriesime rovnicu 
# a najdene riesenia vylucime z definicneho oboru funkcie 
x = var('x')
eq1 = (3*x^2 == 0) 
show(eq1)
eq1sol = solve(eq1, x)
show(eq1sol)
```

2. <ins> Párnosť/nepárnosť </ins>: $\mathbb{D}(f)$ je symetrický vzhľadom na bod $0$; <br/> 
<br/>
$f(-x) = \dfrac{1-2(-x)}{3(-x)^2} = \dfrac{1+2x}{3x^2}$; <br/>
<br/>
pre $x = 1$ je $f(-1) = \dfrac{1+2\cdot 1}{3\cdot 1^2} = \dfrac{3}{3} = 1$; <br/> 
<br/>
pre $x = 1$ je $f(1) = \dfrac{1-2\cdot 1}{3\cdot 1^2} = -\dfrac{1}{3}$. <br/>

5. <ins> Prvá derivácia </ins>: $f'(x) = \dfrac{1}{3}\left ( \dfrac{1-2x}{x^2} \right )' = \dfrac{1}{3}\cdot\dfrac{(1-2x)'\cdot x^2 - (1-2x)\cdot(x^2)'}{(x^2)^2} = \dfrac{1}{3}\cdot \dfrac{(-2)x^2 - (1-2x)\cdot(2x)}{x^4} = \dfrac{1}{3}\cdot\dfrac{2x(-x-(1-2x))}{x^4}$ = <font color=red> $\dfrac{2(x-1)}{3x^3}$ </font>.

#### <font color=blue> Príklad 13 </font> 

Keď je šíp vystrelený do vzduchu, jeho dosah, $R$, je definovaný ako vodorovná vzdialenosť od lukostrelca k bodu, kde šíp dopadne na zem. Ak je zem vodorovná a zanedbáme odpor vzduchu, dá sa ukázať, že platí 

$$R=\dfrac{v_0^2\sin\,(2\theta)}{g},$$  

kde $v_0$ je počiatočná rýchlosť šípu, $g$ je (konštantné) zrýchlenie vďaka gravitácii a $\theta$ je uhol, pod ktorým je šíp vystrelený (viď obrázok), teda $0\le\theta\le\pi/2$. Pod akým uhlom musí byť šíp vystrelený z luku, aby sa maximalizoval dostrel resp. dosah $R$? 


$$\text{Zdroj: Hallet (2017).}$$

<ins> Riešenie: </ins> 

Maximum funkcie $R(\theta)$ je možné nájsť aj bež pužitia kalkulu. Konkrétne funkcia $R$ nadobúda maximum keď $\sin\,(2\theta)=1$ odkiaľ máme riešenie $\theta=\frac{\mathrm{arcsin}\,(1)}{2}=\frac{\pi}{4}$ a hodnota maxima funkcie $R$ v tomto bode je $R(\pi/4)=v_0^2/g$. 

Pozrime sa ako je možné túto úlohu vyriešiť pomocou kalkulu. Potrebujeme nájsť globálne maximum funkcie $R$ pre $0\le\theta\le\pi/2$. Nájdeme stacionárne body funkcie $R$, lebo to sú kandidáti na extrémy funkcie. K tomu potrebujeme prvú deriváciu funkcie $R$ podľa premennej $\theta$: 

$$R'(\theta)=\dfrac{\mbox{d}R}{\mbox{d}\theta}=2\dfrac{v_0^2\cos\,(2\theta)}{g}.$$

Položíme deriváciu rovnú nule, čím dostávame: 

$$2\dfrac{v_0^2\cos\,(2\theta)}{g}=0 \qquad \Leftrightarrow \qquad \cos\,(2\theta)=0 \qquad \Leftrightarrow \qquad 2\theta=\pm\frac{\pi}{2},\pm\frac{3\pi}{2},\pm\frac{5\pi}{2},\ldots,$$ 

teda $\theta=\frac{\pi}{4}$ je jediný stacionárny bod z intervalu $\langle 0, \pi/2 \rangle$. Dostrel pod uhlom $\theta=\pi/4$ je $R(\pi/4)=v_0^2/g$. <font color=orange> Využite Tvrdenie 5 a pre úplnosť overte, že v bode $\theta=\frac{\pi}{4}$ je naozaj lokálne maximum. </font>  <br/> 
Avšak extrém (minimum alebo maximum) funkcie môže byť i v krajnom bode intervalu, preto musíme preveriť funkčné hodnoty v bodoch $\theta=0$ a $\theta=\pi/2$. Ak budú hodnoty v týchto krajných bodoch menšie ako v bode $\theta=\frac{\pi}{4}$, potom môžeme prehlásiť, že v bode $\theta=\frac{\pi}{4}$ je globálne maximum, keďže iný kandidáti na (lokálne) extrémy už nie sú. Všimnime si, že $R=0$ pre obidva krajné body, čo odpovedá tomu, že šíp je vystrelený horizontálne alebo vertikálne. Tým pádom stacionárny bod $\theta=\frac{\pi}{4}$ je lokálne, no zároveň aj globálne maximum funkcie $R$ na intervale $0\le\theta\le\pi/2$. Inými slovami, šíp doletí najďalej, keď je vystrelený z luku pod uhlom $\pi/4$ resp. $45^\circ$.

$$M=2\pi r^2 + 2\pi rh.$$ 

Avšak výška $h$ nie je nezávislá od polomeru $r$ (pozri úvahu na začiatku príkladu a obrázky (a), (b) ). Totiž ak sa $r$ zväčšuje tak súčasne sa $h$ zmenšuje a vice versa (a naopak). Aby sme našli vzťah medzi $r$ a $h$, využijeme fakt, že objem valca $\pi r^2h$ je podľa zadania príkladu rovný hodnote $40\,in^3$ t.j. $\pi r^2h=40$. Inými slovami, 

$$\text{objem plechovky } = \pi r^2h=40, \qquad \text{ odkiaľ } h=\frac{40}{\pi r^2}.$$ 

Toto vyjadrenie pre výšku valca (plechovky) môžeme dosadiť do predošlého vzťahu, konkrétne 

$$\text{materiál na plášť } = 2\pi r h = 2\pi r \frac{40}{\pi r^2}=\frac{80}{r}.$$ 

Dostali sme tak vzorec pre celkový materiál $M$, potrebný na jednu plechovku s polomerom $r$, ktorá má mať objem $40\,in^3$: 

$$M=M(r)=2\pi r^2+\frac{80}{r}.$$ 

Definičný obor tejto funkcie tvoria všetky $r>0$ lebo polomer plechovky nemôže byť záporný ani nulový. <br/>
Aby sme našli minimum funkcie $M$, preskúmame najprv stacionárne body ako kandidátov na hľadané (globálne) minimum. 

$$M'=\frac{\mbox{d}M}{\mbox{d}r}=4\pi r-\frac{80}{r^2}=0 \text{ v kritickom bode, takže } 4\pi r = \frac{80}{r^2}.$$ 

Odtiaľ úpravou ďalej dostávame 

$$\pi r^3=20, \text{ a teda } r=\left(\frac{20}{\pi}\right)^{1/3}=1.85 \text{ palcov},$$ 

čo je v zhode s grafom uvedeným vyššie. Po dosadení do vyjadrenia pre výšku $h$ máme  

$$h=\frac{40}{\pi r^2}=\frac{40}{\pi(1.85)^2}=3.7 \text{ palcov}.$$ 

Materiál potrebný pre jednu plechovku je $M=2\pi(1.85)^2+80/1.85=64.7\, in^2$. 

<font color=orange> Aplikáciou Tvrdenia 5 pre úplnosť overte, že v bode $r=1.85$ je lokálne minimum. </font> 

Aby sme ukázali, že sme naozaj našli globálne (nie iba lokálne) minimum funkcie $M$, pozrime sa na vzorec derivácie $M'$ uvedený vyššie. Pre malé $r$, člen $-80/r^2$ prevažuje (dominuje) a pre veľké $r$ je dominantný člen $4\pi r$. Takže derivácia $M'$ je záporná pre $r<1.85$ a kladná pre $r>1.85$. To znamená (podľa Tvrdenia 2), že funkcia $M$ je klesajúca pre $r<1.85$ a rastúca pre $r>1.85$, preto v bode $r=1.85$ je globálne minimum funkcie $M$.

#### Praktické rady pre riešenie optimalizačných úloh pomocou diferenciálneho kalkulu

1. Uistite sa, že viete, akú veličinu alebo funkciu je potrebné optimalizovať.


2. Ak je to možné, vytvorte niekoľko náčrtov, ktoré ukazujú, ako navzájom sa ovplyvňujúce prvky (veličiny) súvisia. Svoje náčrty jasne označte priradením premenných k veličinám, ktoré sa menia. 


3. Pokúste sa získať vyjadrenie (vzorec/predpis) pre funkciu, ktorá sa má optimalizovať vzhľadom na premenné, ktoré ste identifikovali v predchádzajúcom kroku. Ak je to potrebné, z tohto vzorca vylúčte všetky okrem jednej premennej. Určte doménu (napr. interval), v ktorej táto premenná nadobúda svoje hodnoty. 


4. Nájdite kritické body a výpočtom funkčných hodnôt v týchto bodoch a koncových bodoch (ak sú relevantné) nájdite globálne maximá a ∕ alebo minimá.
