V rámci kurzu MTIb sme si už ukázali, že pre nájdenie miery/intezity zmeny funkcie (miera zmeny vzdialenosti) slúži derivácia (vzdialenosti  podľa času). Konkrétne teda miera zmeny vzdialenosti podľa času t.j. derivácia vzdialenosti podľa času je rýchlosť. Tu si vysvetlíme ako postupovať v opačnom prípade. Teda ak poznáme mieru zmeny funkcie (poznáme rýchlosť = miera  zmeny vzdialenosti podľa času), ako nájdeme pôvodnú funkciu (prejdenú vzdialenosť) ? Začneme tým, ako zistiť prejdenú vzdialenosť (za určitý čas), ak poznáme zmenu vzdialenosti vzhľadom na čas, teda ak poznáme rýchlosť. V prípade kladnej a konštantnej rýchlosti platí nasledovné 

$$\text{vzdialenosť = rýchlosť } \times \text{ čas}.$$ 

V rámci tejto témy si ukážeme, ako odhadnúť a následne presne určiť prejdenú vzdialenosť ak rýchlosť nie je konštantná.

Ako ďaleko auto došlo? Akú vzdialenosť prešlo? Pretože nevieme, ako rýchlo sa auto pohybuje v každom okamihu (nepoznáme funkciu okamžitej rýchlosti - teda nepoznáme jej predpis, iba vidíme graf t.j. modrá krivka), nemôžeme presne vypočítať prejdenú vzdialenosť, ale môžeme urobiť odhad. Rýchlosť sa zvyšuje, takže auto ide prvé dve sekundy rýchlosťou najmenej $20$ ft/s. Pretože vzdialenosť = rýchlosť $\times$ čas, auto prejde počas prvých dvoch sekúnd najmenej $20\cdot2 = 40$  stôp. Podobne to bude počas nasledujúcich dvoch sekúnd najmenej $30\cdot2 = 60$ stôp atď. Počas desaťsekundového časového intervalu to bude najmenej $20\cdot2 + 30\cdot2 + 38\cdot2 + 44\cdot2 + 48\cdot2 = 360$ stôp.

$360$ stôp je teda podhodnoteným odhadom celkovej vzdialenosti prejdenej počas desiatich sekúnd. Aby sme dostali nadhodnotený odhad celkovej vzdialenosti urazenej za 10 sekúnd, môžeme uvažovať podobným spôsobom: Počas prvých dvoch sekúnd je rýchlosť vozidla nanajvýš rovná $30$ ft/s, takže počas prvých dvoch sekúnd prejde nanajvýš $30\cdot2 = 60$ stôp. V nasledujúcich dvoch sekundách prejde najviac ak $38\cdot2 = 76$ stôp atď. Preto počas desaťsekundového intervalu prejde nanajvýš $30\cdot2 + 38\cdot2 + 44\cdot2 + 48\cdot2 + 50\cdot2 = 420$ stôp.

Keď sa zamyslíme nad odobma odhadmi cekovej vzdialenosti prejdenej autom počas $10$ sekúnd, tak dostávame nasledovný vzťah 

$$360 \text{ ft } \le \text{ celková prejdená vzdialenosť } \le 420 \text{ ft }.$$

Rozdiel medzi dolným a horným odhadom je $60$ stôp. Lepší odhad by mohol byť priemer týchto dvoch odhadov t.j. $\dfrac{360+420}{2}=390$ stôp. 

Horný aj dolný odhad môžeme znázorniť na grafe rýchlosti v závislosti od času. Rýchlosť možno graficky znázorniť nanesením údajov z predchádzajúcej tabuľky v podobe bodov a nakreslením krivky cez tieto body (pozri nasledujúci obrázok) Plocha prvého tmavého obdĺžnika je $20\cdot2 = 40$, čo je dolný odhad vzdialenosti prejdenej počas prvých dvoch sekúnd. Plocha druhého tmavého obdĺžnika je $30\cdot2 = 60$, dolný odhad vzdialenosti urazenej počas nasledujúcich dvoch sekund. Celková plocha tmavých obdĺžnikov predstavuje podhodnotený odhad celkovej vzdialenosti prejdenej počas desiatich sekúnd. Ak uvažujeme tmavé a svetlé obdĺžniky spolu dokopy t.j. ku každému tmavému obdĺžniku pridáme svetlý obĺžnik nad ním, potom je plocha prvého takto "zlepeného" obdĺžnika rovná hodnote $30\cdot2 = 60$, čo je vlastne horný (nadhodnotený) odhad vzdialenosti prejdenej za prvé dve sekundy. Plocha druhého skombinovaného obdĺžnika (tmavý plus svetlý obdĺžnik) je $38\cdot2 = 76$, t.j. horný odhad pre dráhu prejdenú počas ďalších dvoch sekúnd. Pokračovanie tohto výpočtu naznačuje, že horný odhad celkovej vzdialenosti prejdenej autom počas desiatich sekúnd predstavuje súčet plôch tmavých a svetlých obdĺžnikov. 


$$\text{Zdroj: Hallet (2014).}$$

Keď robíme častejšie merania rýchlosti, obdĺžniky použité na odhad prejdenej vzdialenosti lepšie priliehajú ku krivke (viď nasledujúce tri obrázky) a teda odhady celkovej prejdenej dráhy za určitý čas sú presnejšie. V limite, keď sa počet delení (časového) intervalu zväčšuje, tzn. meriame rýchosť častejšie, vidíme, že prejdená vzdialenosť je daná obsahom oblasti medzi krivkou (okamžitej) rýchlosti a vodorovnou osou. Vo všeobecnosti teda platí: **Ak je rýchlosť kladná, potom celková prejdená vzdialenosť je rovná obsahu oblasti pod krivkou rýchlosti** (to je vlastne hodnota určitého integrálu z funkcie vyjadrujúcej okamžitú rýchlosť auta v čase $t$ na intervale $\langle 0,10 \rangle$, pričom spomeňme si, že okamžitá rýchlosť je miera zmeny prejdenej dráhy za čas t.j. derivácia dráhy podľa času - viď krivka v grafoch). 

 
$$\text{Zdroj: Hallet (2014).}$$

Keď je rýchlosť pohybujúceho sa hmotného bodu daná funkciou $y=v(t)$ na intervale $\langle a,b \rangle$, potom plocha pod krivkou (danou grafom funkcie) na intervale $\langle a,b \rangle$ udáva celkovú vzdialenosť prejdenú hmotným bodom počas časového intervalu $\langle a,b \rangle$. Plocha pod nejakou krivkou danou funkciou $y=f(x)$ môže byť zaujímavá aj v iných aplikáciách, či bez ohľadu na význam a interpretáciu funkcie $f$. Budeme teraz ďalej pokračovať v približnom určovaní  (aproximácii) obsahu plochy pod krivkou danou grafom určitej funkcie. Túto aproximácu budeme realizovať pomocou rôznych typov obdĺžnikov a ich súčtov. Len pripomíname, že uvažujeme (ak to nie je uvedené ináč) funkcie, ktoré sú spojité na intervale $\langle a,b \rangle$ a teraz pre názornejšiu ilustráciu aj nezáporné. 

Pri uvažovaní o takej aproximácii potrebujeme najprv určiť počet obdĺžnikov $n$, ktorých súčet obsahov bude rozumne aproximovať obsah plochy medzi krivkou danou nejakou funkciou a osou $x$. Pre jednoduchosť budeme uvažovať obdĺžniky, ktoré majú rovnakú šírku a teda rozdelíme interval $\langle a,b \rangle$ na $n$ intervalov rovnakej dĺžky. Potom šírka jedného obdĺžnika bude rovná hodnote $\Delta x=\frac{b-a}{n}$. Ďalej položíme $x_0=a$, $x_n=b$ a definujeme ďalšie body v rámci intervalu $\langle a,b \rangle$ takto $x_i=a+i\Delta x$ pre $i\in\{1,2,\ldots,n-1\}$, pričom $x_1=x_0+\Delta x$, $x_2=x_0+2\Delta x$ atď. Nasledujúci obrázok ilustruje popísanú situáciu. 

 
$$\text{Zdroj: Boelkins et al. (2018).}$$

Každý interval v tvare $\langle x_{i-1},x_i \rangle$ tvorí základňu (šírku) obdĺžnika. Potrebujeme ešte stanoviť výšku takého obdĺžnika. Núka sa niekoľko prirodzených možností (v skutočnosti ich je nekonečne veľa). Štandardne sa zvykne ako výška vziať funkčná hodnota $f(x_{i-1})$ v ľavom krajnom bode intervalu $\langle x_{i-1},x_i \rangle$, alebo funkčná hodnota $f(x_{i})$ v pravom krajnom bode intervalu $\langle x_{i-1},x_i \rangle$ alebo funkčná hodnota v bode, ktorý je stredom intervalu $\langle x_{i-1},x_i \rangle$ t.j. $f\left(\frac{x_{i}+x_{i-1}}{2}\right)$. 

Uvažujme ľubovoľnú kladnú funkciu na intervale $\langle a,b \rangle$. Taktiež predpokladajme, že máme interval $\langle a,b \rangle$ rozdelený tak, ako je to znázornené a popísané vyššie. Zvoľme teraz výšky obdĺžnikov ako funkčné hodnoty v ľavých krajných bodoch t.j. $f(x_{i-1})$. Potom obsah obdĺžnika prislúchajúceho intervalu $\langle x_{i-1},x_i \rangle$ je daný vzťahom $A_{i}=f(x_{i-1})\cdot\Delta x$, znázorňuje to nasledujúci obrázok. 


$$\text{Zdroj: Boelkins et al. (2018).}$$

Označme $L_n$ celkovú zelenú plochu na obrázku pozostávajúcu z odĺžnikov, zrejme platí 

$$L_n=A_1+A_2+\ldots+A_{i-1}+\ldots+A_n=f(x_0)\cdot\Delta x+f(x_1)\cdot\Delta x+\ldots+f(x_i)\cdot\Delta x+\ldots+f(x_{n-1})\cdot\Delta x=\sum\limits_{i=1}^{n}f(x_{i-1})\Delta x.$$

Výraz pre $L_n$ sa zvykne nazývať **_ľavý Riemannov súčet_** funkcie $f$ na intervale $\langle a,b \rangle$. 

Analogicky je možné zaviesť _pravé_ a _stredné_ _Riemannové_ súčty. Situácie sú znázornené na ďalšom obrázku. 

 
$$\text{Zdroj: Boelkins et al. (2018).}$$

Ak zvolíme za výšku obdĺžnika na intervale $\langle x_{i-1},x_i \rangle$ hodnotu $f(x_i)$ (predchádzajúci obrázok vľavo), potom obsah takého obdĺžnika je daný vzťahom $B_i=f(x_i)\cdot\Delta x$. Súčet obsahov všetkých týchto obdĺžnov označíme $R_n$ a je daný vzťahom 

$$R_n=B_1+B_2+\ldots+B_{i}+\ldots+B_n=f(x_1)\cdot\Delta x+f(x_2)\cdot\Delta x+\ldots+f(x_{i})\cdot\Delta x+f(x_n)\cdot\Delta x=\sum\limits_{i=1}^{n}f(x_{i})\Delta x.$$ 

Výraz pre $R_n$ sa zvykne nazývať **_pravý Riemannov súčet_** funkcie $f$ na intervale $\langle a,b \rangle$. 

Body, ktoré ležia v strede intervalov $\langle x_{i-1},x_i \rangle$ označme $t_i=\dfrac{x_{i-1}+x_i}{2}$. Ako posledný prípad uvažujme obdĺžniky, ktorých výška je daná hodnotou $f(t_i)$. Napr. prvý obdĺžnik  na predchádzajúcom obrázku vpravo má obsah $C_1=f(t_1)\cdot\Delta x$. Preto súčet vsetkých takto vytvorených obdĺžkov je rovný 

$$M_n=C_1+C_2+\ldots+C_i+\ldots C_n=f(t_1)\cdot\Delta x+f(t_2)\cdot\Delta x+\ldots+f(t_i)\cdot\Delta x+\ldots+f(t_n)\cdot\Delta x=\sum\limits_{i=1}^{n}f(t_{i})\Delta x.$$ 

Výraz pre $M_n$ sa zvykne nazývať **_stredný Riemannov súčet_** funkcie $f$ na intervale $\langle a,b \rangle$. 

**Ak $f(x)\ge0$ pre $x\in \langle a,b \rangle$, tak každý z Riemannových súčtov $L_n$, $R_n$, $M_n$ poskytuje odhad obsahu plochy  pod krivkou $y=f(x)$ na intervale $\langle a,b \rangle$.** V kontexte nezápornej rýchlosti danej funkciou $y=v(t)$ odpovedajúce Riemannové súčty aproximujú vzdialenosť prejdenú hmotným bodom s rýchlosťou $v$ počas časového intervalu $\langle a,b \rangle$. 

**O Riemannových súčtoch je možné uvažovať aj všeobecnejšie napr. v tom zmysle, že výšky obdĺznikov môžu odpovedať hodnotám $f(t_i)$, pričom $t_i$ je ľubovoľný bod z intervalu $\langle x_{i-1},x_i \rangle$  (nie nutne ľavý/pravý krajný bod intervalu alebo prostredný bod) t.j. $x_{i-1}\le t_i \le x_{i}$. Hovorí o tom Definícia 2.** 

---

### <font color=red> Definícia 1 </font>  

Nech $\langle a,b \rangle$ je uzavretý interval. **_Označené delenie_** intervalu $\langle a,b \rangle$ je postupnosť $(x_0,t_1,x_1, t_2, x_2,\dots, x_{n-1}, t_n, x_n)$ reálnych čísel taká, že $a = x_0 \leq t_1 \leq x_1 \leq t_2 \leq x_2 \leq \dots \leq x_{n-1} \leq t_n \leq x_n = b$. 

---

---

#### Poznámka 2

Označené delenie teda možno chápať ako rozdelenie základného intervalu $\langle a,b \rangle$ na $n$ podintervalov $\langle x_{i-1},x_i \rangle$, pričom každému z nich je priradený nejaký bod $t_i \in \langle x_{i-1},x_i \rangle$ ("značka" intervalu) - viď ľavé alebo práve krajné body intervalov $\langle x_{i-1},x_i \rangle$, prípadne stredy intervalov $\langle x_{i-1},x_i \rangle$ (pozri úvahy vyššie).

--- 

--- 

### <font color=red> Definícia 2 </font> 

Nech $f$ je funkcia definovaná na uzavretom intervale $\langle a,b \rangle$ a nech $P = (x_0,t_1,x_1, t_2, x_2,\dots, x_{n-1}, t_n, x_n)$ je nejaké označené delenie tohto intervalu. **_Riemannov súčet_** $S_n$ funkcie $f$ vzhľadom k deleniu $P$ je súčet

$$S_n = \sum\limits_{i = 1}^n f(t_i)\Delta x_i$$
	
kde $\Delta x_i = x_i - x_{i-1}$ je dĺžka intervalu $\langle x_{i-1},x_i \rangle$.

---

---

#### Poznámka 3

* Všimnime si, že podľa Definície 2 môžeme vyberať body $t_i$ napr. ako stredy intervalov $\langle x_{i-1}, x_{i} \rangle$, alebo ako ľavý či pravý krajný bod intervalu $\langle x_{i-1}, x_{i} \rangle$, prípadne iný bod patriaci do intervalu $\langle x_{i-1}, x_{i} \rangle$. 


* Taktiež si uvedomme, že dĺžky intervalov $\Delta x_i=x_i-x_{i-1}$ v zvolenom delení $P = (x_0,t_1,x_1, t_2, x_2,\dots, x_{n-1}, t_n, x_n)$ intervalu $\langle a,b \rangle$ nemusia mať všetky rovnakú dĺžku (viď Príklad 6 podľa Madaras, 2020). Ak by sme vzali rovnomerné delenie intervalu, potom by platilo $\Delta x_i=\Delta x=\frac{b-a}{n}$ a teda každý interval v delení by mal rovnakú dĺžku rovnú hodnote $\frac{b-a}{n}$. Pre jednoduchosť budeme uvažovať predovšetkým rovnomerné delenia intervalu $\langle a,b \rangle$. 


* $L_n$, $R_n$, $M_n$ sú špeciálne prípady Riemannových súčtov $S_n$, kde $\Delta x_i=\Delta x=\frac{b-a}{n}$ a body $t_i$ sú vyberané tiež špeciálnym spôsobom. 


* Je zrejmé, že ak uvažujeme dve (pre jednoduchosť) rovnomerné delenia $P_1$, $P_2$ intervalu $\langle a,b \rangle$, pričom delenie $P_1$ má deliace body $x_0,t_1,x_1,\ldots,x_{m-1},t_m,x_m$ a delenie $P_2$ má deliace body $x_0,t_1,x_1,\ldots,x_{n-1},t_n,x_n$ kde $m<n$, tak potom dĺžka $\Delta x$ podintervalov $\langle x_{i-1}, x_{i} \rangle$ v delení $P_1$ je väčšia ako v delení $P_2$ (viac deliacich bodov $\to$ viac podintervalov $\to$ kratšie podintervaly). 


* Všimnime si aj to, že ak budeme pridávať deliace body (zjemňovať delenie intervalu $\langle a,b \rangle$), tzn. budeme deliť interval $\langle a,b \rangle$ na stále viac podintervalov tzn. $n\to\infty$ (vznikne tak viac obdĺžnikov), potom dĺžka jednotlivých podintervalov (šírka obdĺžnikov) v delení sa bude zmenšovať a pre $n\to\infty$ bude $\Delta x \to 0$. 


* **Ak funkcia $f$ nie je vždy kladná na  $\langle a,b \rangle$, tak Riemannove súčty funkcie vzhľadom k označeným deleniam zodpovedajú celkovému signovanému obsahu obdĺžnikov** (pričom obsahy obdĺžnikov, ktoré sú pod osou $x$, sa zarátavajú do Riemannovho súčtu resp. obsahu so záporným znamienkom)

Využite interaktívny kód  na preskúmanie Riemannových súčtov $L_n$. K akej hodnote sa blíži $L_n$ so zväčšujúcim sa $n$? 

```python
# Riemann Sum (with left endpoints) and Sage Interact - an example
# input : integrand f; boundary points of interval <a, b>; n - the number of rectangles in Riemann sum
# output : plot of f and the area between the graph of f and the x axis approximated by the rectangles of a Riemann sum (with left endpoints)
# author: P. Szabó, 16 March 2018 
# modified: A. Gajdoš (2020)
@interact
def LeftRiemannSum( 
    f = input_box(default = x^3-3*x^2+2*x) ,
    a = input_box(default =  0) ,
    b = input_box(default = 1.5) ,
    n = slider(vmin = 1, vmax = 20, default = 8, step_size=1)  ): 
    LSn_val = var('Sn_val')
    LSn_val = 0 
    f(x)=f
    #plot the function
    p = plot(f, x, a,b, color = "red", thickness = 4, \
             alpha = 0.5, fillcolor = "blue", fill = true)
    dx = (b-a)/n
    #plot each trapezoid
    for i in [1,2,..,n]:
    #for the shading ...
        xbegin = a+(i-1)*dx
        xend   = a+i*dx
        p += polygon( [(xbegin,0),  (xend,0),               \
                       (xend, f(xbegin)), (xbegin,f(xbegin))],  \
                       color="yellow",alpha = 0.5, aspect_ratio=
                       "automatic")
       #these are optional for a better graph:
        p += line([(xend,0), (xend, f(xbegin))], color = "black")
        p += line([(xend, f(xbegin)), (xbegin,f(xbegin))], color = "black")
        p += line([(xbegin, f(xbegin)),(xbegin,0) ], color = "black")
        LSn_val += f(xbegin)*(xend-xbegin)
    p.show(title = "Riemann Sum with Left Endpoints", ticks = dx) 
    
    show("Riemann Sum with Left Endpoints")
    LSn=LatexExpr("L_n = \\sum_{i=1}^{n}f(t_i)\\Delta x=")
    show(LSn+latex(LSn_val)+LatexExpr("\\doteq")+latex(N(LSn_val)))
```

#### Poznámka 4

V predošlých (interaktívnych) príkladoch sme videli, že keď sa zväčšuje počet obdĺžnikov, tzn. zväčšuje sa $n$ v Riemannových  súčtoch (pridávame deliace body t.j. zjemňuje sa delenie intervalu $\langle a,b \rangle$ a zmenšuje sa $\Delta x$), potom hodnoty Riemannových súčtov $L_n$, $M_n$ a $R_n$ stále presnejšie aproximujú signovanú plochu medzi grafom funkcie a osou $x$, navyše sa hodnoty týchto Riemannových súčtov blížia k rovnakej hodnote (pre rastúce $n$). Ukazuje sa, že tento jav nastáva pre ľubovoľnú spojitú funkciu na intervale $\langle a,b \rangle$ a pre Riemannov súčet $S_n$, v ktorom je $t_i$ ľubovoľný bod z intervalu $\langle x_{i-1},x_i \rangle$ prislúchajúceho deleniu $P$. Preto ak necháme $n\to\infty$, nebude záležať na tom, kde budeme počítať funkčnú hodnotu $f(t_i)$ v rámci daného podintervalu príslušného delenia, lebo platí 

$$\lim_{n\to\infty}L_n=\lim_{n\to\infty}R_n=\lim_{n\to\infty}M_n=\lim_{n\to\infty}\sum\limits_{i=1}^{n}f(t_i)\Delta x.$$

### <font color=red> Definícia 3 </font>

**_Riemannov určitý integrál_** spojitej funkcie $f$ na uzavretom intervale $\langle a,b \rangle$ je reálne číslo ozn. $\displaystyle\int_{a}^{b}f(x)\,\mathrm{d}x$ dané vzťahom 

$$\displaystyle\int_{a}^{b}f(x)\,\mathrm{d}x=\lim_{n\to\infty}\sum\limits_{i=1}^{n}f(t_i)\Delta x,$$

kde $\Delta x=\frac{b-a}{n}$, $x_i=a+i\Delta x$ pre $i\in\{0,1,\ldots,n\}$ a $x_{i-1} \le t_i \le x_i$ pre $i\in\{1,\ldots,n\}$, číslo $a$ sa nazýva **_dolná hranica_**, číslo $b$ **_horná hranica_**. Symbol $\mathrm{d}x$ vyjadruje **integračnú premennú** (súvisí to s $\Delta x$). 

--- 

--- 

#### Poznámka 5 

* Ináč povedané, ak si spomenieme na definíciu limity postupnosti, tak môžeme povedať, že Riemannov určitý integrál spojitej funkcie $f$ na uzavretom intervale  $\langle a,b \rangle$  je reálne číslo $s=\lim\limits_{n\to\infty}S_n=\lim\limits_{n\to\infty}\sum\limits_{i=1}^{n}f(t_i)\Delta x$ také, pre ktoré platí: </br>
$$\left(\forall\varepsilon>0\right)\left(\exists\, n_0\in\mathbb{N}\right)\left(\forall n\ge n_0\right)\quad \left|\sum\limits_{i=1}^{n}f(t_i)\Delta x-s\right|<\varepsilon;$$ <br/>
</br>
pričom $S_n$ je nejaký Riemannov súčet, pre jednoduchosť ale uvažujme rovnomerné delenie intervalu $\langle a,b \rangle$.  


* Postupnosť Riemannových súčtov $\{S_n\}$ je vlastne číselná postupnosť, ktorej členy vyjadrujú aproximáciu (danú súčtom obsahov obdĺňikov) signovaného obsahu medzi krivkou $y=f(x)$ a osou $x$ (viď interaktívne príklady). S rastúcim $n$ t.j. pre $n\to\infty$ sú aproximácie stále presnejšie. Samozrejme miesto $S_n$ si môžeme predstaviť napr. konkrétne aj $L_n$, $R_n$, $M_n$ alebo iné. 


* Geometrická interpretácia Riemannovho integrálu: číslo $s$ je rovné skutočnému signovanému obsahu plochy medzi grafom funkcie $f$, osou $x$, priamkami $x=a,x=b$. Definícia potom hovorí, že pre ľubovoľne danú presnosť $\varepsilon$, s ktorou aproximujeme skutočný signovaný obsah $s$ pomocou sumy $\sum\limits_{i = 1}^n f(t_i)\Delta x$, existuje prirodzené číslo $n_0$ také, že pre $n\ge n_0$, Riemannov súčet každého označeného (rovnomerného) delenia s $n$ obdĺžnikmi je v medziach daných zvolenou presnosťou  t.j. $ \left |\sum\limits_{i = 1}^n f(t_i)\Delta x - s  \right | < \varepsilon$.

### <font color=red> Definícia 4 </font>

Funkcia $f$ sa nazýva (Riemannovsky) **_integrovateľná_** na uzavretom intervale $\langle a,b\rangle$, ak existuje Riemannov integrál $f$ na $\langle a,b\rangle$. 

---

#### Poznámka 6

* Keďže úvahy o Riemannovom určitom integrále je možne robiť aj všeobecnejšie a je možné ho zaviesť (definovať) aj pre iné funkcie (pre všeobecnejšiu - väčšiu triedu funkcií, do ktorej patria i spojité funkcie), tak podľa Definície 4 rozumieme integrovateľnou funkciou ne intervale $\langle a,b \rangle$ takú (nie nutne spojitú) funkciu $f$, ktorej určitý integrál na intervale $\langle a,b \rangle$ je nejaké reálne číslo (nie $\pm\infty$). Geometricky to teda znamená, že plocha resp. signovaný obsah medzi grafom funkcie a osou $x$ je konečné reálne číslo. 


* Veta 1 hovorí, že u funkcie, ktorá je spojitá na uzavretom intervale nenastávajú problémy ohľadne existencie určitého integrálu na tomto intervale, a že je to vždy konečné reálne číslo (konečná signovaná plocha medzi grafom funkcie a osou $x$), čiže taká funkcia je vždy integrovateľná na uzavretom intervale. Táto veta nám vlastne garantuje korektnosť definície určitého integrálu pre spojitú funkciu na uzavretom intervale $\langle a,b \rangle$ (definujeme objekt - určitý (Riemannov) integrál, ktorý existuje a pre spojitú funkciu na uzavretom intervale $\langle a,b \rangle$ má zmysel).

#### Poznámka 7

Riemannov integrál funkcie na uzavretom intervale možno chápať aj ako zobrazenie, ktoré dvojici čísel $a,b$ (hraniciam uzavretého intervalu) a funkcii $f$ priradí hodnotu $\displaystyle\int\limits_a^b f(x)\,\mathrm{d}x$. Je vhodné rozšíriť definíciu aj pre prípad, keď $a > b$ (čo nezodpovedá situácii v bežnom intervale) zavedením konvencie

$$
\displaystyle\int\limits_a^b f(x)\,\mathrm{d}x = -\displaystyle\int\limits_b^a f(x)\, \mathrm{d}x.
$$

Pre $a=b$ z toho vyplýva

$$
\displaystyle\int\limits_a^a f(x)\,\mathrm{d}x = 0.
$$

Geometricky to znamená, že ak vieme vypočítať obsah plochy (sivá) napr. pod krivkou určenou funkciou $f(t)=3e^{-3t}$ na intervale $\langle 1.1,1.6 \rangle$ t.j. $\displaystyle\int_{1.1}^{1.6}3e^{-3t}\,\mathrm{d}t$, tak potom vieme vypočítať aj obsah plochy (žltá) pod rovnakou krivkou ale na menšom intervale obsiahnutom v $\langle 1.1,1.6 \rangle$, napr. na podintervale $\langle 1.2,1.4 \rangle$ t.j. $\displaystyle\int_{1.2}^{1.4}3e^{-3t}\,\mathrm{d}t$. No a prirodzene bude platiť medzi plochami nerovnosť $\displaystyle\int_{1.2}^{1.4}3e^{-3t}\,\mathrm{d}t \le \displaystyle\int_{1.1}^{1.6}3e^{-3t}\,\mathrm{d}t$, viď žltá plocha je menšia ako sivá plocha.

**Nasledujúca veta nám dáva návod, ako vypočítať určitý integrál zo spojitej funkcie $f$ na intervale $\langle a,b \rangle$, ak poznáme primitívnu funkciu $F$ k funkcii $f$.** Tým pádom vieme využiť poznatky o neurčitých integráloch z predošlých dvoch tém. Inými slovami, ak nájdeme primitívnu funkciu $F$ k funkcii $f$ - t.j. vypočítame najprv neurčitý integrál $\displaystyle\int f(x)\,\mathrm{d}x$ a položíme $F(x)=\displaystyle\int f(x)\,\mathrm{d}x$, potom už len stačí urobiť rozdiel funkčných hodnôt $F(b)-F(a)$, kde $a,b$ sú krajné body intervalu $\langle a,b \rangle$, cez ktorý integrujeme a položiť $\displaystyle\int\limits_a^b f(x)\,\mathrm{d}x=F(b)-F(a)$. 

Všimnime si, že teraz - pri výpočte určitého integrálu $\displaystyle\int\limits_a^b f(x)\,\mathrm{d}x$, keď hľadáme primitívnu funkciu $F$ k funkcii $f$, nepotrebujeme brať do úvahy integračnú konštantu $C$, lebo ak by sme miesto $F(x)$ vzali $F(x)+C$, potom $C$ vypadne, keďže $F(b)+C-\left(F(a)+C\right)=F(b)+C-F(a)-C=F(b)-F(a)$.

### <font color=green> Veta 2 (Newtonova-Leibnizova) </font> 

Nech funkcia $f$ je na intervale $\langle a,b\rangle$ spojitá a nech $F$ je funkcia primitívna
k $f$. Potom

$$\displaystyle\int\limits_a^b f(x)\,\mathrm{d}x = F(b)-F(a).$$

<font color=purple> EXTRA: Dôkaz. <br/> 
Predpokladajme, že $f$ je nezáporná na $\langle a,b\rangle$ (všeobecný prípad sa odvodí podobne). Pre $t > a$ nech $U(a,t)$ je útvar ohraničený grafom funkcie $f$, osou $x$ a priamkami $x=a,x=t$ a nech $S(a,t)$ označuje jeho plošný obsah (ako funkcia $t$). Potom pre malé hodnoty $h>0$ sa útvary $U(a,t+h)$ a $U(a,t)$ líšia len o "obdĺžnik" so stranami $h$ a $f(t)$:

 
$$\text{Zdroj: Madaras (2020).}$$

Potom možno približne písať

$$
S(a,t+h)-S(a,t) \doteq f(t)\cdot h
$$

z čoho vyplýva

$$
\dfrac{S(a,t+h)-S(a,t)}{h} \doteq f(t)
$$

Ak budeme teraz uvažovať limitu pre $h \to 0$, tak dostávame $S'(a,t) = f(t)$. 

Teda $S(a,t)$ je primitívna funkcia ku funkcii $f$ a keďže (podľa interpretácie Riemannovho integrálu ako plošného obsahu) $S(a,t) = \displaystyle\int\limits_a^t f(x)\, \mathrm{d}x$, tak $\displaystyle\int\limits_a^t f(x)\, \mathrm{d}x$ je primitívna funkcia ku $f$.

Nech $F$ je nejaká primitívna funkcia ku $f$. Keďže ľubovoľné dve primitívne funkcie sa líšia len o konštantu, tak pre všetky $t \in \langle a,b \rangle$ platí $F(t) - \displaystyle\int\limits_a^t f(x)\, \mathrm{d}x = c$ pre pevné $c \in \mathbb{R}$. 

Táto rovnosť platí aj pre $t = a$ a $t=b$, čo dáva

$$
F(a) - \displaystyle\int\limits_a^a f(x)\, \mathrm{d}x = c
$$

$$
F(b) - \displaystyle\int\limits_a^b f(x)\, \mathrm{d}x = c
$$

z čoho porovnaním pravých strán rovníc máme

$$
F(a) - 0 = F(b) - \displaystyle\int\limits_a^b f(x)\, \mathrm{d}x
$$

teda

$$F(b) - F(a)  =  \displaystyle\int\limits_a^b f(x)\, \mathrm{d}x.$$

</font>

#### <font color=blue> Príklad 9 </font>

Metódou per partes vypočítajte $\displaystyle\int\limits_{2}^{3}\ln x\,\mathrm{d}x$. 

<ins> Riešenie: <ins/>
   
Aby sme mohli aplikovať metódu per partes, potrebujeme si uvedomiť, že platí $\ln x=1\cdot \ln x$ a teda 

$$\displaystyle\int\limits_{2}^{3}\ln x\,\mathrm{d}x=\displaystyle\int\limits_{2}^{3}1\cdot \ln x\,\mathrm{d}x.$$ 

Ďalej položme $v=\ln x$, odkiaľ dostaneme $v'=1/x$ a ešte položme $u'=1$ odkiaľ $u=x$. Po aplikovaní metódy per partes máme 

$$\displaystyle\int\limits_{2}^{3}\underbrace{\ln x}_{v} \cdot \underbrace{1}_{u'}\,\mathrm{d}x=\left[\underbrace{\ln x}_{v}\cdot \underbrace{x}_{u}\right]_{2}^{3}-\displaystyle\int\limits_{2}^{3}\underbrace{\frac{1}{x}}_{v'}\cdot \underbrace{x}_{u}\,\mathrm{d}x=\left[x\ln x\right]_{2}^{3}-\displaystyle\int\limits_{2}^{3}1\,\mathrm{d}x=\left[x\ln x\right]_{2}^{3}-\left[x\right]_{2}^{3}=(3\ln3-2\ln2)-(3-2)=3\ln3-2\ln2-1.$$

Všimnime si, že pri aplikácii metódy per partes na určitý integrál, potrebujeme dosadiť integračné hranice $2,3$ jednak pre súčin $u\cdot v$ t.j. $\left[x\cdot\ln x\right]_{2}^{3}$ ale aj v integráli $\displaystyle\int u'\cdot v\,\mathrm{d}x$ t.j. $\displaystyle\int\limits_{2}^{3} \frac{1}{x}\cdot x\,\mathrm{d}x$.

#### <font color=blue> Príklad 10 </font>

Pomocou substitúcie vypočítajte $\displaystyle\int\limits_{0}^{2}xe^{x^2}\,\mathrm{d}x$. 

<ins> Riešenie: <ins/>

Ukážeme si dva prístupy ako sa správne vysporiadať s hranicami určitého integrálu. 

Vypočítame najprv neurčitý integrál. Urobme substitúciu $w=x^2$, potom $\mathrm{d}w=2x\,\mathrm{d}x$ odkiaľ $\frac{1}{2}\mathrm{d}w=x\,\mathrm{d}x$. Takto dostávame 

$$\displaystyle\int xe^{x^2}\,\mathrm{d}x=\displaystyle\int e^{w}\frac{1}{2}\,\mathrm{d}w=\frac{1}{2}e^{w}+C=\frac{1}{2}e^{x^2}+C.$$

Teraz už môžeme využiť Newtonovu-Leibnizovu formulu 

$$\displaystyle\int\limits_{0}^{2}xe^{x^2}\,\mathrm{d}x=\left[\frac{1}{2}e^{x^2}\right]_{0}^{2}=\frac{1}{2}e^{2^2}-\frac{1}{2}e^{0^2}=\frac{1}{2}(e^4-e^0)=\frac{1}{2}(e^4-1).$$

Mohli sme však postupovať aj ináč. Pri substitúcii sme mohli hneď vyjadriť, ako sa vďaka použitej substitúcii zmenia hranice určitého integrálu. Konkrétne teda, keďže sme položili $w=x^2$ a $x$ nadobúda hodnody v pôvodonom integrále od $0$ do $2$, tak potom $w$ bude nadobúdať hodnoty v novom integrále medzi hranicami $0^2$ a $2^2$ t.j. od $0$ po $4$. Jednoducho stačí do vyjadrenia $w=x^2$ dosadiť hranice pre $x$, t.j. $x=0$, $x=2$ a získame tak nové hranice pre $w$. Integrál potom vyzerá nasledovne 

$$\displaystyle\int\limits_{0}^{2}xe^{x^2}\,\mathrm{d}x=\displaystyle\int\limits_{x=0}^{x=2}xe^{x^2}\,\mathrm{d}x=\displaystyle\int\limits_{w=0}^{w=4}e^{w}\frac{1}{2}\,\mathrm{d}w=\displaystyle\int\limits_{0}^{4}e^{w}\frac{1}{2}\,\mathrm{d}w=\left[\frac{1}{2}e^w\right]_{0}^{4}=\frac{1}{2}e^4-\frac{1}{2}e^0=\frac{1}{2}(e^4-1).$$

Dostali sme rovnaký výsledok ako v predchádzajúcom prípade. 

Všimnime si ešte, že ak sme ponechali pôvodné hranice určitého integrálu (1. spôsob), museli sme sa v substitúcii vrátiť k pôvodnej premennej $x$ a následne dopočítať určitý integrál pomocou Newtonovej-Leibnizovej formuly. Ak sme určili pri substitúcii nové hranice, už sme sa nevracali k pôvodnej premennej $x$ a určitý integrál sme dopočítali pomocou Newtonovej-Leibnizovej formuly s novou premennou $w$.

#### Poznámka 8 

* Podobne ako Leibnizova notácia $\dfrac{\mathrm{d}y}{\mathrm{d}x}$ v prípade derivácii nám pripomína, že derivácia je vlastne limita pomeru (podielu) diferencií resp. prírastkov $\dfrac{\Delta y}{\Delta x}$, kde $\Delta x$ reprezentuje malú zmenu v premennej $x$, tak aj notácia integrálu $\displaystyle\int_{a}^{b}f(x)\,\mathrm{d}x$ nám pripomína, že integrál je limita súčtu (Riemannové súčty). Keďže na $\mathrm{d}x$ je možné nahliadať ako na malú zmenu (prírastok) v premennej $x$ t.j. $\Delta x$, potom v integrálnych (Riemannových) súčtoch vystupujú súčiny v tvare "$f(x)$ krát prírastok v $x$" t.j. $f(x)\cdot \Delta x$. To nám dáva nasledujúcu interpretáciu: **Jednotky (merania) pre výsedok integrálu $\displaystyle\int_{a}^{b}f(x)\,\mathrm{d}x$ sú dané súčinom jednotiek pre $f(x)$ a $x$.** 


* Príklady: Ak $x$ a $f(x)$ sú v rovnakých jednotkách, napr. v cm, potom výsledok integrálu $\displaystyle\int_{a}^{b}f(x)\,\mathrm{d}x$ je v $\mathrm{cm^2}$ a reprezentuje obsah/plochu; ak $v=f(t)$ je rýchlosť v m/s a $t$ je čas v sekundách, potom výsledok integrálu $\displaystyle\int_{a}^{b}f(t)\,\mathrm{d}t$ je v $\mathrm{(m/s)\cdot s=m}$ a reprezentuje zmenu polohy, teda prejdenú dráhu (vzdialenosť); ak $C(t)$  vyjadruje cenu tepla za vykurovanie domu za jeden deň v dolároch a $t$ je čas meraný v dňoch, potom $\displaystyle\int_{0}^{90}C(t)\,\mathrm{d}t$ reprezentuje poplatok za teplo za vykurovanie domu počas 90 dní a výsledok  integrálu $\displaystyle\int_{0}^{90}C(t)\,\mathrm{d}t$ je v dolároch lebo $\mathrm{doláre/deň\cdot deň=doláre}$

b) Zvýšila sa alebo poklesla hodnota investície v priebehu piatich mesiacov?

<ins> Riešenie: <ins/>

a) Hodnota investície sa znížila počas prvých troch mesiacov, pretože miera zmeny hodnoty investície je vtedy záporná (derivácia záporná). Hodnota investície vzrástla za posledné 2 mesiace, pretože miera zmeny hodnoty investície je vtedy kladná  (derivácia kladná). 

b) Chceme nájsť celkovú zmenu hodnoty investície medzi $t = 0$ a $t = 5$. Keďže celková zmena hodnoty akcie je rovná integrálu z miery (intenzity) zmeny hodnoty investície $F'(t)$, zaujíma nás teda 

$$\text{Celková zmena hodnoty akcie } = \displaystyle\int^{5}_{0}F'(t)\,\mathrm{d}t.$$

Integrál sa rovná obsahu oblasti nad osou $t$ (vyznačenej na obrázku) mínus obsah oblasti pod osou $t$ (vyznačenej na obrázku). Pretože na obrázku je plocha pod osou väčšia ako plocha nad osou, integrál je záporný. Celková zmena hodnoty investície v tomto päťmesačnom období je záporná, preto sa hodnota investície za päť mesiacov znížila.
