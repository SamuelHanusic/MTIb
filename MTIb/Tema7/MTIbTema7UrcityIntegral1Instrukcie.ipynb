<a id=motivacia></a>
# <font color=brown> Motivácia</font>


 

V predchádzajúcej téme sme sa venovali hľadaniu primitívnych funkcií resp. výpočtu tzv. neurčitého integrálu. Nadobudnuté vedomosti a zručnosti využijeme teraz pri výpočte tzv. **určitých integrálov**. 

 

Najprv zavedieme **pojem určitého integrálu**, priblížime si jeho **geometrickú interpretáciu**, ktorá bude okrem iného dôležitá i neskôr pri jeho **aplikáciách**. Pozrieme sa na to, ako vypočítať hodnoty rôznych typov **určitých integrálov** pomocou **Newton-Leibnizovej formuly** a tiež pomocou metódy **per partes** i **substitúcie**, mierne upravených pre **určitý integrál**. Na kokrétnych príkladoch si ukážeme, ako je možné **interpretovať hodnotu určitého integrálu** v kontexte danej úlohy a tiež preštudujeme **základne vlastnosti určitého integrálu**.

# <font color=brown> Učebné ciele</font>

### Základné 


* Rozumieť definícii Riemanovho (určitého) integrálu, jeho geometrickej interpretácii a pojmom, ktoré s tým súvisia (postupnosť delení intervalu, norma delenia,  Riemanov súčet). 


* Vedieť, kedy je funkcia integrovateľná (tvrdenie). 


* Vysloviť vlastnosti (linearita) Riemanovho integrálu a ilustrovať na príkladoch (aj geometricky). 


* Rozumieť Newtonovej-Leibnizovej formule pre výpočet určitého integrálu a ilustrovať jej použitie na príklade. 


* Rozumieť metóde per partes a substitučnej metóde pre výpočet určitých integrálov. 

### Pokročilé 


* Aplikovať Newton-Leibnizovu formulu pri výpočte určitých integrálov. 


* Využiť techniky výpočtu neurčitých integrálov pri výpočte určitých integrálov vybraných základných typov.

1. Rýchlosť bicykla je daná funkciou $v(t)=5t$, kde $t$ je čas v sekundách a $v$ je rýchlosť v stopách za sekundu. Akú dráhu prejde bicykel za $3$ sekundy? Nakreslite aj obrázok a pri výpočte sa inšpirujte myšlienkami z Príkladu 1. 


2. Pre funkciu $3x^2 + 2x -3$ určte pomocou interaktívnych kódov uvedených vyššie hodnoty Riemannových súčtov $L_n$, $R_n$, $M_n$ pre $n\in\{5,10,15,20\}$. Čo sa deje s hodnotami $L_n$, $R_n$, $M_n$, keď sa počet intervalov $n$ zväčšuje?  


3. Nájdite dvojicu funkcií $f$, $g$, pre ktorú bude platiť na konkrétnom intervale $\langle a,b \rangle$ vlastnosť: Ak $f(x) \leq g(x)$, potom $\displaystyle\int\limits_a^b f(x)\, \mathrm{d}x \leq \displaystyle\int\limits_a^b g(x)\, \mathrm{d}x$. Napíšte konkrétne funkcie $f,g$, konkrétny interval $\langle a,b \rangle$ a vykreslite aj grafy nájdených funkcií na zvolenom intervale (do jednéh obrźka). 


4. Pomocou Newtonovej-Leibnizovej formuly vypočítajte $\displaystyle\int\limits_{\pi/6}^{3\pi/2} 4\sin(2x)\,\mathrm{d}x$ a výpočet overte v Sage.
